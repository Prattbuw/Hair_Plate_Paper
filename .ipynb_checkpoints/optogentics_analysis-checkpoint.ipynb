### Optogenetic Silencing

Characterize how silencing CoHP8 neurons alters the body-coxa angles (L1_A flex/rot/abduct) during walking and turning

Assess how silencing CoHP8 neurons alters the distrbution of joint angles during grooming

```python
def add_stim_col(df):
    df['stim_on_off'] = 0
    df.loc[(df['stimlen'] == 1) & (df['fnum'] >= 150) & (df['fnum'] < 450), 'stim_on_off'] = 1
    return df
```

```python
# new transforms - abduction and flexion are swapped and rotation angle is already corrected for
def transform_rotation_angle(df):
    # transform coordinate system into Chris'
    df['L1A_rot'] = df['L1A_rot'].values - 90

    return df

def transform_flexion_angle(df):

    # covert to full polar from half polar representation
    df['L1A_abduct'][df['L1A_abduct'] > 180] = df['L1A_abduct'][df['L1A_abduct'] > 180].values - 360

    # transform coordinate system into Chris'
    idxs_1 = np.logical_and(df['L1A_abduct'] >= 0, df['L1A_abduct'] <= 180) 
    idxs_2 = np.logical_and(df['L1A_abduct'] < 0, df['L1A_abduct'] >= -180) 
    df['L1A_abduct'][idxs_1] = 180 - df['L1A_abduct'][idxs_1].values 
    df['L1A_abduct'][idxs_2] = df['L1A_abduct'][idxs_2].values + 180
    
    return df
```

```python
from tqdm import tqdm
```

```python
'''Extract walking steps not dependent on walking behavioral classifier'''
def optogenetic_silencing_results(df):

    # isolate walking bouts
    fps = 300 # 300 frames/second
    walking_bout_threshold = 0.2 * fps # walking bout must be greater than 200 ms
    N_interp_samples = 100
    interp_phases = np.linspace(0,1,N_interp_samples)
    '''Change to this'''
    vr_threshold = 25 # deg/s was 3 mm/s - rotational velocticity threshold # grant used 10 mm/s
#     vr_threshold = 3 # deg/s was 3 mm/s - rotational velocticity threshold # grant used 10 mm/s
    vf_min_threshold = 5 # mm/s - minimum forward velocity threshold - z
    vf_max_threshold = 25 # mm/s - maximum forward velocity threshold - increase if needed - y
    vslip_threshold = 5 # mm/s - maximum amount of side slip velocity threshold - x
    array_state = 0

    t1_grooming_data_stim_on = []
    t1_grooming_data_stim_off = []

    antennal_grooming_data_stim_on = []
    antennal_grooming_data_stim_off = []

    head_eye_grooming_data_stim_on = []
    head_eye_grooming_data_stim_off = []

    # iterate through flies
    for fly in tqdm(np.unique(df['flyid']).tolist()):
        t1_grooming_abduct_on_data = np.array([])
        t1_grooming_abduct_off_data = np.array([])
        t1_grooming_rot_on_data = np.array([])
        t1_grooming_rot_off_data = np.array([])
        t1_grooming_flex_on_data = np.array([])
        t1_grooming_flex_off_data = np.array([])

        ant_grooming_abduct_on_data = np.array([])
        ant_grooming_abduct_off_data = np.array([])
        ant_grooming_rot_on_data = np.array([])
        ant_grooming_rot_off_data = np.array([])
        ant_grooming_flex_on_data = np.array([])
        ant_grooming_flex_off_data = np.array([])

        he_grooming_abduct_on_data = np.array([])
        he_grooming_abduct_off_data = np.array([])
        he_grooming_rot_on_data = np.array([])
        he_grooming_rot_off_data = np.array([])
        he_grooming_flex_on_data = np.array([])
        he_grooming_flex_off_data = np.array([])

        for condition in np.unique(df['cond']).tolist(): 
            for repeat in np.unique(df['rep']).tolist(): # 10 repeats for each condition
                trial_data = df[(df['flyid'] == fly)&(df['rep'] == repeat)&(df['cond'] == condition)]
                # if laser is off, don't need to worry about when a walking or grooming bout occurs
                bout_data = trial_data

                # next filter for L1 swing stance - find the non-filtered period
                if len(bout_data[~np.isnan(bout_data['L1_swing_stance'].values)]) > 0:
                    swing_stance_data = bout_data[~np.isnan(bout_data['L1_swing_stance'].values)]

                    # extract out steps - -1 is swing to stance # currently swing starts
                    stance_starts = np.where(np.diff(swing_stance_data['L1_swing_stance'].values) == -1)[0]+1

                    if len(stance_starts)> 1:
                        # iterate through steps and isolate out L1_joint angle data
                        for step in range(len(stance_starts)-1):
                            
                            # basic filters for steps: mean forward velocity and step duration
                            step_duration = (stance_starts[step+1] - stance_starts[step])/fps
                            
                            '''fictrac_delta_rot_lab_y_mms_raw for split-half control'''
#                             vf = np.mean(swing_stance_data['fictrac_delta_rot_lab_y_mms_raw'][stance_starts[step]:stance_starts[step+1]].values)
                            vf = np.mean(swing_stance_data['fictrac_delta_rot_lab_y_mms'][stance_starts[step]:stance_starts[step+1]].values)
                            
                            if (vf > vf_min_threshold) and (step_duration >= 0.05):
                                L1A_flex = swing_stance_data['L1A_abduct'][stance_starts[step]:stance_starts[step+1]].values
                                L1A_abduct = swing_stance_data['L1A_flex'][stance_starts[step]:stance_starts[step+1]].values
                                L1A_rot = swing_stance_data['L1A_rot'][stance_starts[step]:stance_starts[step+1]].values
 
                                # interpolate angles so that they are all aligned in phase
                                curr_phases = np.linspace(0,1,len(L1A_flex))
                                f_flex = interpolate.interp1d(curr_phases, L1A_flex)
                                f_abduct = interpolate.interp1d(curr_phases, L1A_abduct)
                                f_rot = interpolate.interp1d(curr_phases, L1A_rot)

                                L1A_flex_interp = f_flex(interp_phases)
                                L1A_abduct_interp = f_abduct(interp_phases)
                                L1A_rot_interp = f_rot(interp_phases)

                                # determine conversion into mm
                                L1A_x = swing_stance_data['L1A_x'][stance_starts[step]:stance_starts[step+1]].values
                                L1A_y = swing_stance_data['L1A_y'][stance_starts[step]:stance_starts[step+1]].values
                                L1A_z = swing_stance_data['L1A_z'][stance_starts[step]:stance_starts[step+1]].values
                                L1B_x = swing_stance_data['L1B_x'][stance_starts[step]:stance_starts[step+1]].values
                                L1B_y = swing_stance_data['L1B_y'][stance_starts[step]:stance_starts[step+1]].values
                                L1B_z = swing_stance_data['L1B_z'][stance_starts[step]:stance_starts[step+1]].values

                                L_px = np.mean(np.sqrt((L1B_x-L1A_x)**2 + (L1B_y-L1A_y)**2 + (L1B_z-L1A_z)**2))
                                conversion_mm = 0.4/L_px # average coxa length is 0.25mm for males - igor measured 0.4 mm

                                # average trajectory of the distal tip of the coxa
                                coxa_x = (L1B_x - L1A_x)*conversion_mm
                                coxa_y = (L1B_y - L1A_y)*conversion_mm
                                coxa_z = (L1B_z - L1A_z)*conversion_mm

                                f_coxa_x = interpolate.interp1d(curr_phases,  coxa_x)
                                f_coxa_y = interpolate.interp1d(curr_phases,  coxa_y)
                                f_coxa_z = interpolate.interp1d(curr_phases,  coxa_z)

                                coxa_x_interp = f_coxa_x(interp_phases)
                                coxa_y_interp = f_coxa_y(interp_phases)
                                coxa_z_interp = f_coxa_z(interp_phases)


                                # extract 3D postion of the tarsi tips
                                L1E_x = swing_stance_data['L1E_x'][stance_starts[step]:stance_starts[step+1]].values*conversion_mm
                                L1E_y  = swing_stance_data['L1E_y'][stance_starts[step]:stance_starts[step+1]].values*conversion_mm
                                L1E_z  = swing_stance_data['L1E_z'][stance_starts[step]:stance_starts[step+1]].values*conversion_mm

                                # phase of tarsi position
                                f_x = interpolate.interp1d(curr_phases, L1E_x)
                                f_y = interpolate.interp1d(curr_phases, L1E_y)
                                f_z = interpolate.interp1d(curr_phases, L1E_z)

                                L1E_x_interp = f_x(interp_phases)
                                L1E_y_interp = f_y(interp_phases)
                                L1E_z_interp = f_z(interp_phases)

                                # classify the step as being a forward walking or turning step and stimulation state
                                '''use fictrac_delta_rot_lab_z_deg/s_raw for split-half control'''
#                                 vr = np.mean(swing_stance_data['fictrac_delta_rot_lab_z_deg/s_raw'][stance_starts[step]:stance_starts[step+1]].values)
                                vr = np.mean(swing_stance_data['fictrac_delta_rot_lab_z_deg/s'][stance_starts[step]:stance_starts[step+1]].values)
                                step_stim = swing_stance_data['stim_on_off'][stance_starts[step]:stance_starts[step+1]].values
                                swing_start = (np.where(np.diff(swing_stance_data['L1_swing_stance'][stance_starts[step]:stance_starts[step+1]].values)==1)[0][0]+1)/len(swing_stance_data['L1_swing_stance'][stance_starts[step]:stance_starts[step+1]])

                                # Determine stimulation state
                                if np.all(step_stim==0): # stim off
                                    stim_state = 0
                                elif np.all(step_stim==1): # stim on
                                    stim_state = 1
                                else:
                                    stim_state = 2 # transition stim state

                                # determine walking state
                                # forward walking
                                if (vf > vf_min_threshold) and (np.abs(vr)<vr_threshold):
                                    walking_state = 0 # forward walking
                                elif (vr<-vr_threshold) and (vf > vf_min_threshold): # elif (vr<-vr_threshold) and (vf > vf_min_threshold):
                                    walking_state = 1 # left turn (negative in deg/s) - positive ball rotation is the fly turning left
                                elif (vr>vr_threshold) and (vf > vf_min_threshold): #(vr>vr_threshold) and (vf > vf_min_threshold): 
                                    walking_state = 2 # right turn (positive)
                                else:
                                    walking_state = 3 # some other walking state

                                # concatenate data 
                                if array_state == 0:
                                    array_state = 1

                                    # create meta data matrix
                                    meta_data = np.array([fly, walking_state, stim_state, vf, vr, swing_start]) # fly, walking state, stim state, forward velocity, rotational velocity, swing start
                                    rot_data = L1A_rot_interp
                                    flex_data = L1A_flex_interp
                                    abduct_data = L1A_abduct_interp
                                    L1x_data = L1E_x_interp
                                    L1y_data = L1E_y_interp
                                    L1z_data = L1E_z_interp
                                    coxa_x_data = coxa_x_interp 
                                    coxa_y_data = coxa_y_interp
                                    coxa_z_data = coxa_z_interp


                                else:
                                    meta_data = np.vstack((meta_data, np.array([fly, walking_state, stim_state, vf, vr, swing_start])))
                                    rot_data = np.vstack((rot_data, L1A_rot_interp))
                                    flex_data = np.vstack((flex_data, L1A_flex_interp))
                                    abduct_data = np.vstack((abduct_data, L1A_abduct_interp))
                                    L1x_data = np.vstack((L1x_data, L1E_x_interp))
                                    L1y_data = np.vstack((L1y_data, L1E_y_interp))
                                    L1z_data = np.vstack((L1z_data, L1E_z_interp))
                                    coxa_x_data = np.vstack((coxa_x_data, coxa_x_interp))
                                    coxa_y_data = np.vstack((coxa_y_data, coxa_y_interp))
                                    coxa_z_data = np.vstack((coxa_z_data, coxa_z_interp))


                # Grooming analysis
                # t1_grooming
                t1_grooming_bout_data = trial_data[~np.isnan(trial_data['t1_grooming_bout_number'].values)]
                if len(t1_grooming_bout_data)>0: # see if grooming bouts are detected
 
                    L1A_flex_t1_grooming_stim_off = t1_grooming_bout_data['L1A_abduct'][t1_grooming_bout_data['stim_on_off'].values ==0].values
                    L1A_flex_t1_grooming_stim_on = t1_grooming_bout_data['L1A_abduct'][t1_grooming_bout_data['stim_on_off'].values ==1].values
                    t1_grooming_flex_off_data = np.concatenate((t1_grooming_flex_off_data, L1A_flex_t1_grooming_stim_off))
                    t1_grooming_flex_on_data = np.concatenate((t1_grooming_flex_on_data, L1A_flex_t1_grooming_stim_on))
       
                    L1A_abduct_t1_grooming_stim_off = t1_grooming_bout_data['L1A_flex'][t1_grooming_bout_data['stim_on_off'].values ==0].values
                    L1A_abduct_t1_grooming_stim_on = t1_grooming_bout_data['L1A_flex'][t1_grooming_bout_data['stim_on_off'].values ==1].values
                    t1_grooming_abduct_on_data = np.concatenate((t1_grooming_abduct_on_data, L1A_abduct_t1_grooming_stim_on))
                    t1_grooming_abduct_off_data = np.concatenate((t1_grooming_abduct_off_data,L1A_abduct_t1_grooming_stim_off))

                    L1A_rot_t1_grooming_stim_off = t1_grooming_bout_data['L1A_rot'][t1_grooming_bout_data['stim_on_off'].values ==0].values
                    L1A_rot_t1_grooming_stim_on = t1_grooming_bout_data['L1A_rot'][t1_grooming_bout_data['stim_on_off'].values ==1].values
                    t1_grooming_rot_on_data = np.concatenate((t1_grooming_rot_on_data, L1A_rot_t1_grooming_stim_on))
                    t1_grooming_rot_off_data = np.concatenate((t1_grooming_rot_off_data, L1A_rot_t1_grooming_stim_off))


                # antennal grooming
                ant_grooming_bout_data = trial_data[~np.isnan(trial_data['antennae_grooming_bout_number'].values)]
                if len(ant_grooming_bout_data)>0: # see if grooming bouts are detected
                    # isolate the different angles for each of the bouts
                    L1A_flex_ant_grooming_stim_off = ant_grooming_bout_data['L1A_flex'][ant_grooming_bout_data['stim_on_off'].values ==0].values
                    L1A_flex_ant_grooming_stim_on = ant_grooming_bout_data['L1A_flex'][ant_grooming_bout_data['stim_on_off'].values ==1].values
                    ant_grooming_flex_off_data = np.concatenate((ant_grooming_flex_off_data, L1A_flex_ant_grooming_stim_off))
                    ant_grooming_flex_on_data = np.concatenate((ant_grooming_flex_on_data, L1A_flex_ant_grooming_stim_on))


                    L1A_abduct_ant_grooming_stim_off = ant_grooming_bout_data['L1A_abduct'][ant_grooming_bout_data['stim_on_off'].values ==0].values
                    L1A_abduct_ant_grooming_stim_on = ant_grooming_bout_data['L1A_abduct'][ant_grooming_bout_data['stim_on_off'].values ==1].values
                    ant_grooming_abduct_on_data = np.concatenate((ant_grooming_abduct_on_data, L1A_abduct_ant_grooming_stim_on))
                    ant_grooming_abduct_off_data = np.concatenate((ant_grooming_abduct_off_data,L1A_abduct_ant_grooming_stim_off))

                    L1A_rot_ant_grooming_stim_off = ant_grooming_bout_data['L1A_rot'][ant_grooming_bout_data['stim_on_off'].values ==0].values
                    L1A_rot_ant_grooming_stim_on = ant_grooming_bout_data['L1A_rot'][ant_grooming_bout_data['stim_on_off'].values ==1].values
                    ant_grooming_rot_on_data = np.concatenate((ant_grooming_rot_on_data, L1A_rot_ant_grooming_stim_on))
                    ant_grooming_rot_off_data = np.concatenate((ant_grooming_rot_off_data, L1A_rot_ant_grooming_stim_off))


        # Save grooming data for each fly
        t1_grooming_data_stim_on.append([t1_grooming_flex_on_data, t1_grooming_abduct_on_data, t1_grooming_rot_on_data])
        t1_grooming_data_stim_off.append([t1_grooming_flex_off_data, t1_grooming_abduct_off_data, t1_grooming_rot_off_data])

        antennal_grooming_data_stim_on.append([ant_grooming_flex_on_data, ant_grooming_abduct_on_data, ant_grooming_rot_on_data])
        antennal_grooming_data_stim_off.append([ant_grooming_flex_off_data, ant_grooming_abduct_off_data, ant_grooming_rot_off_data])

    return meta_data, rot_data, flex_data, abduct_data, L1x_data, L1y_data, L1z_data, t1_grooming_data_stim_on, t1_grooming_data_stim_off, antennal_grooming_data_stim_on, antennal_grooming_data_stim_off,coxa_x_data, coxa_y_data, coxa_z_data  
                                            
                                    
```

```python
def bootstrap_stats(n_bootstraps, pval_thresh, bon_correction, pop1, pop2, plotting, comparasion):
    print(comparasion)
    n_pop1 = len(pop1)
    n_pop2 = len(pop2)
    mean_pop_diff = np.mean(pop2) - np.mean(pop1) # BL
    pooled_pop = np.concatenate((pop1, pop2))

    # bootstraping 
    mean_stat_diff = np.zeros(n_bootstraps)
    for j in range(n_bootstraps):
        resample_pop1 = np.random.permutation(pooled_pop)[0:n_pop1] # randomly permute from a uniform distribution
        resample_pop2 = np.random.permutation(pooled_pop)[n_pop1::]
        mean_stat_diff[j] = np.mean(resample_pop2) - np.mean(resample_pop1)

    if plotting:
        plt.figure(1, figsize =[5,5])
        hist, bin_edges = np.histogram(mean_stat_diff, bins=30, density=True)
        plt.hist(mean_stat_diff, bins = 30, color='k', density = True)
        plt.plot(np.array([mean_pop_diff, mean_pop_diff]), np.array([0,max(hist)]), color ='r')

    # compute p-value
    if mean_pop_diff> 0:
        n_sig_samples = len(np.where(mean_stat_diff>=mean_pop_diff)[0])
    else:
        n_sig_samples = len(np.where(mean_stat_diff<=mean_pop_diff)[0])
    pvalue = n_sig_samples/n_bootstraps
    if pvalue == 0: # correct for the number of bootstraps
        pvalue = 1/n_bootstraps
    print('p value = ', pvalue)

    # significance threshold with bonferroni correction
    if pvalue <= (pval_thresh/bon_correction):
        print('Statistically Significant')
    else:
        print('Not Statistically Significant')
```

```python
def transform_rotation_angle(df):
    # covert to full polar from half polar representation
    df['L1A_rot'][df['L1A_rot'] < 0] = 360 + df['L1A_rot'][df['L1A_rot'] < 0].values

    # transform coordinate system into Chris'
    idxs_1 = df['L1A_rot'] >= 90
    idxs_2 = df['L1A_rot'] < 90
    df['L1A_rot'][idxs_1] = df['L1A_rot'][idxs_1].values - 90
    df['L1A_rot'][idxs_2] = df['L1A_rot'][idxs_2].values +270
    
    return df

# convert flexion angle from Grant's setup to Chris'
# df = filt_crtl_sil_df.copy()
def transform_flexion_angle(df):

    # covert to full polar from half polar representation
    df['L1A_flex'][df['L1A_flex'] > 180] = df['L1A_flex'][df['L1A_flex'] > 180].values - 360

    # transform coordinate system into Chris'
    idxs_1 = np.logical_and(df['L1A_flex'] >= 0, df['L1A_flex'] <= 180) 
    idxs_2 = np.logical_and(df['L1A_flex'] < 0, df['L1A_flex'] >= -180) 
    df['L1A_flex'][idxs_1] = 180 - df['L1A_flex'][idxs_1].values 
    df['L1A_flex'][idxs_2] = df['L1A_flex'][idxs_2].values + 180
    
    return df
```

```python
def optogenetic_activation_results(df, prestim_duration, poststim_duration):
    fps = 300 # 300 frames/second
    dt = 1/fps
    prestim_window = int(prestim_duration * fps) # walking bout must be greater than 250ms before and after stimulation 
    poststim_window = int(poststim_duration * fps) # walking bout must be greater than 250ms before and after stimulation
    vf_threshold = 5 # mm/s - minimum forward velocity threshold - z
    tarsi_vel_threshold = 5 #mm/s
    array_state = 0

    # iterate through flies
    for fly in np.unique(df['flyid']).tolist():
        for condition in np.unique(df['cond']).tolist(): # 1 is laser off and 2 is laser on; was cond
            for repeat in np.unique(df['rep']).tolist(): # 10 repeats for each condition
                trial_data = df[(df['flyid'] == fly)&(df['rep'] == repeat)&(df['cond'] == condition)]
                
                if len(trial_data) == 600:
                    # determine if fly is stationary prior to stim onset
                    vf = trial_data['fictrac_delta_rot_lab_y_mms'].values[150-prestim_window:150]
                    
                    # determine conversion into mm
                    L1A_x = trial_data['L1A_x'].values
                    L1A_y = trial_data['L1A_y'].values
                    L1A_z = trial_data['L1A_z'].values
                    L1B_x = trial_data['L1B_x'].values
                    L1B_y = trial_data['L1B_y'].values
                    L1B_z = trial_data['L1B_z'].values

                    L_px = np.mean(np.sqrt((L1B_x-L1A_x)**2 + (L1B_y-L1A_y)**2 + (L1B_z-L1A_z)**2))
                    conversion_mm = 0.4/L_px # average coxa length is 0.25mm for males - igor measured 0.4

                    '''Pre-stim filter'''
                    # extract 3D postion of the tarsi tips
                    L1E_x = trial_data['L1E_x'].values*conversion_mm
                    L1E_y  = trial_data['L1E_y'].values*conversion_mm
                    L1E_z  = trial_data['L1E_z'].values*conversion_mm
                    tarsi_velocity = gaussian_filter1d((np.sqrt(np.diff(L1E_x)**2 + np.diff(L1E_y)**2 + np.diff(L1E_z)**2)/dt),2)[150-prestim_window:150]

                    # standing filter - if true - analyze trial
                    prestim_standing = np.all(~np.isnan(trial_data['standing_bout_number'].values[150-prestim_window:150]))

                    # check that the fly isn't grooming its abdomen or t3 legs
                    prestim_t3_grooming = np.all(np.isnan(trial_data['t3_grooming_bout_number'].values[150-prestim_window:150]))
                    prestim_abdomen_grooming = np.all(np.isnan(trial_data['abdomen_grooming_bout_number'].values[150-prestim_window:150]))

                    # determine prestim state - select for non-moving
                    vf_filt = np.all(np.abs(vf) < vf_threshold)
                    vtip_filt = np.all(tarsi_velocity<tarsi_vel_threshold)
                    prestim_filter = np.all(np.array([vf_filt, vtip_filt, prestim_t3_grooming, prestim_abdomen_grooming]))

                    # post-stim filter - ensure that the stim doesn't result in walking or grooming
                    '''Post-stim filter'''
                    poststim_walking = np.all(np.isnan(trial_data['walking_bout_number'].values[150:150+poststim_window]))
                    poststim_t1_grooming = np.all(np.isnan(trial_data['t1_grooming_bout_number'].values[150:150+poststim_window]))
                    poststim_t3_grooming = np.all(np.isnan(trial_data['t3_grooming_bout_number'].values[150:150+poststim_window]))
                    poststim_abdomen_grooming = np.all(np.isnan(trial_data['abdomen_grooming_bout_number'].values[150:150+poststim_window]))
                    poststim_antennae_grooming = np.all(np.isnan(trial_data['antennae_grooming_bout_number'].values[150:150+poststim_window]))
                    poststim_ball_push = np.all(np.isnan(trial_data['ball_push_bout_number'].values[150:150+poststim_window]))                                    
                    poststim_filter = np.all(np.array([poststim_walking, poststim_t1_grooming, poststim_t3_grooming, poststim_abdomen_grooming, 
                                                      poststim_antennae_grooming, poststim_ball_push]))

                    poststim_standing = np.any(~np.isnan(trial_data['standing_bout_number'].values[150:150+poststim_window]))

                    if (prestim_filter or prestim_standing) and (poststim_filter or poststim_standing): # if true analyze data
                        # extract joint angles during the window
                        L1A_flex = trial_data['L1A_flex'].values[150-prestim_window:150+poststim_window]
                        L1A_abduct= trial_data['L1A_abduct'].values[150-prestim_window:150+poststim_window]
                        L1A_rot= trial_data['L1A_rot'].values[150-prestim_window:150+poststim_window]
#                         L1A_rot[L1A_rot>90] = -180 - (180 - L1A_rot[L1A_rot>90]) # corrects for wrap around

                        # speed
                        v_forward = trial_data['fictrac_delta_rot_lab_y_mms'].values[150-prestim_window:150+poststim_window]
                        v_rot= np.abs(trial_data['fictrac_delta_rot_lab_y_mms'].values[150-prestim_window:150+poststim_window])

                        # concatenate data 
                        if array_state == 0:
                            array_state = 1

                            # create meta data matrix
                            if np.any(trial_data['stim_on_off'].values == 1):
                                stim_state = 1
                            else:
                                stim_state = 0
                            meta_data = np.array([fly, stim_state, repeat]) # fly, stim on/off ,trial
                            rot_data = L1A_rot - L1A_rot[0] # correct for initial condition
                            flex_data = L1A_flex - L1A_flex[0]
                            abduct_data = L1A_abduct - L1A_abduct[0]
                            vf_data = v_forward - np.mean(v_forward)
                            vr_data = v_rot - np.mean(v_rot)
                            L1x_data = L1E_x[150-prestim_window:150+poststim_window] - L1E_x[150-prestim_window:150+poststim_window][0]
                            L1y_data = L1E_y[150-prestim_window:150+poststim_window] - L1E_y[150-prestim_window:150+poststim_window][0]
                            L1z_data = L1E_z[150-prestim_window:150+poststim_window] - L1E_z[150-prestim_window:150+poststim_window][0]
                        else:
                            if np.any(trial_data['stim_on_off'].values == 1):
                                stim_state = 1
                            else:
                                stim_state = 0
                            meta_data = np.vstack((meta_data, np.array([fly, stim_state, repeat])))
                            rot_data = np.vstack((rot_data, L1A_rot - L1A_rot[0]))
                            flex_data = np.vstack((flex_data, L1A_flex - L1A_flex[0]))
                            abduct_data = np.vstack((abduct_data, L1A_abduct - L1A_abduct[0]))
                            vf_data = np.vstack((vf_data, v_forward - np.mean(v_forward)))
                            vr_data = np.vstack((vr_data, v_rot - np.mean(v_rot)))
                            L1x_data = np.vstack((L1x_data, L1E_x[150-prestim_window:150+poststim_window] - L1E_x[150-prestim_window:150+poststim_window][0]))
                            L1y_data = np.vstack((L1y_data, L1E_y[150-prestim_window:150+poststim_window] - L1E_y[150-prestim_window:150+poststim_window][0]))
                            L1z_data = np.vstack((L1z_data, L1E_z[150-prestim_window:150+poststim_window] - L1E_z[150-prestim_window:150+poststim_window][0]))

    return meta_data, rot_data, flex_data, abduct_data, L1x_data, L1y_data, L1z_data, vf_data, vr_data
                                    
```

```python
# process data for control and cohp8 silenced flies
prestim_duration = 0.1
poststim_duration = 0.5 # a good window to minimize behavioral variability
cohp_act_meta_data, cohp_act_rot_data, cohp_act_flex_data, cohp_act_abduct_data, cohp_act_L1x_data, cohp_act_L1y_data, cohp_act_L1z_data, cohp_vf_data, cohp_vr_data = optogenetic_activation_results(filt_cohp_act_df, prestim_duration, poststim_duration)
```
