```python
## isolate example data for each trial
FS = 200
nonsplit = np.where(meta_matrix[1][:,4] != 1)[0]
trials = np.unique(nonsplit)

'''ADJUST THE TRIAL NUMBER HERE'''
trial_num = 133 #149 # THE BEST EXAMPLE
trial = trials[trial_num] 

# data from trial
print('fly: ', meta_matrix[1][trial,0])
print('trial: ', meta_matrix[1][trial,1])
BL = meta_matrix[1][trial,-1]
r1_x = limb_positions[1][0,0,:,trial] * BL
l1_x = limb_positions[1][0,3,:,trial] * BL
behs = beh_class[1][trial,:]
vel = parallel_vel[1][trial,:] * BL
heading = heading_angle[1][trial,:]
sw_st = swing_stance_mat[1][np.array([0,3]), :, trial]
time = np.linspace(0,len(behs)/FS, len(behs))
t1 = 10.25#6.15#time[0] # ex1 (133):7.6 
t2 = 12.25#13.15#time[-1] # ex1 (133): 12.6

# plot example data
fig = plt.figure(1, figsize = [5,5])

# front leg x positions
plt.subplot(4,1,1)
plt.plot(time, r1_x, color = (235/255, 33/255, 39/255), linewidth = 1.5)
plt.plot(time, l1_x, color = (29/255, 118/255, 187/255), linewidth = 1.5)

'''Add behaviors'''
beh_colors = ['salmon', 'gray']
cnt =-1
for beh in [1, 4]:
    cnt +=1
    idxs = np.where(behs == beh)[0]
    diff_idxs = np.diff(idxs)
    transitions = np.where(diff_idxs>1)[0]
    if (len(idxs) > 0) and (len(transitions) == 0): # if there is just one long bout
            plt.axvspan(time[idxs[0]], time[idxs[-1]], ymin=0, ymax=2, color = beh_colors[cnt], alpha = 0.5)
    if len(transitions) >0:
        if len(transitions) == 1: # just 2 bout:
            plt.axvspan(time[idxs[0]], time[idxs[transitions[0]]], ymin=0, ymax=2, color = beh_colors[cnt], alpha = 0.5)
            plt.axvspan(time[idxs[transitions[0]+1]], time[idxs[-1]], ymin=0, ymax=2, color = beh_colors[cnt], alpha = 0.5)
        else: # multiple bouts
            start_idx = idxs[0]
            for j in range(len(transitions)):
                if len(transitions)-1 == j:
                    end_idx = idxs[transitions[j]]
                    plt.axvspan(time[start_idx], time[end_idx], ymin=0, ymax=2, color = beh_colors[cnt], alpha = 0.5)
                    plt.axvspan(time[idxs[transitions[j]+1]], time[idxs[-1]], ymin=0, ymax=2, color = beh_colors[cnt], alpha = 0.5)
                else:
                    end_idx = idxs[transitions[j]]
                    plt.axvspan(time[start_idx], time[end_idx], ymin=0, ymax=2, color = beh_colors[cnt], alpha = 0.5)
                    start_idx = idxs[transitions[j]+1]
                    
# axes
plt.ylabel('x pos (mm)', fontsize = 12, fontname = 'arial')
plt.yticks(fontsize = 10, fontname = 'arial')
plt.xticks([])
plt.xlim(t1, t2)
plt.ylim([0,2.5])

# swing stance plot
plt.subplot(4,1,2)
labels = ['R1','L1']
axes = plt.gca()
axes.set_yticks(np.arange(0, 2, 1))
axes.set_yticklabels(labels)
plt.imshow(sw_st, interpolation = 'none', cmap = 'gray',aspect='auto')
plt.ylabel('leg', fontsize = 12, fontname = 'arial')
plt.yticks(fontsize = 10, fontname = 'arial')
plt.xticks([])
plt.xlim(t1*200, t2*200)

# heading angle
plt.subplot(4,1,3)
plt.plot(time, heading, color = 'black', linewidth = 3.5)
plt.ylabel('heading angle (deg)', fontsize = 12, fontname = 'arial')
plt.yticks(fontsize = 10, fontname = 'arial')
plt.ylim([-60,60])
plt.xticks([])
plt.xlim(t1, t2)

# forward velocity
plt.subplot(4,1,4)
plt.plot(time, vel, color = 'black', linewidth = 3.5)
plt.ylabel('Velocity (mm/s)', fontsize = 12, fontname = 'arial')
plt.xlabel('Time (s)', fontsize = 12, fontname = 'arial')
plt.yticks(fontsize = 10, fontname = 'arial')
plt.xticks(fontsize = 10, fontname = 'arial')
plt.ylim([-5,35])
plt.xlim(t1, t2)

plt.tight_layout()
plt.show

fig.savefig(figure_dir + 'treadmill_ex_3.png', dpi = 300)
fig.savefig(figure_dir + 'treadmill_ex_3.svg', dpi = 300)
```

### Posture Analysis - isolate frames during standing

```python
def posture_distributions_stats(metric, meta_matrix, beh_class, sign_clause, zero_filter, neg_ones_filter):
    stat_store = []
    dist_store = []

    # iterate through genotypes
    for gen in range(len(genotypes)):
        nflies = len(np.unique(meta_matrix[gen][:, 0]).astype(int).tolist()) # how many flies

        # create stat matrix
        stat_matrix = np.zeros((nflies, 5))
        fly_dist = []

        # iterate through flies
        for fly in range(nflies):
            #isolate trials of that fly
            fly_idxs = meta_matrix[gen][:, 0] == fly+1
            sb_idxs = meta_matrix[gen][:, 4] != 1 # just presplit and post split
            trial_idxs = np.logical_and.reduce((fly_idxs, sb_idxs))
            fly_data = sign_clause * metric[gen][trial_idxs,:].flatten() # flatten data
            fly_beh_class = beh_class[gen][trial_idxs,:].flatten() 

            # filter on behavior class
            filter_fly_data = fly_data[fly_beh_class==4]
            filter_fly_data = filter_fly_data[~np.isnan(filter_fly_data)] # filter out nans

            # additional filter
            if zero_filter:
                filter_fly_data = filter_fly_data[filter_fly_data != 0] 

            if neg_ones_filter:
                filter_fly_data = filter_fly_data[filter_fly_data != -1]

            # store data of fly-specific metric distrbution and stats
            fly_dist.append(filter_fly_data)
            stat_matrix[fly, 0] = np.mean(filter_fly_data) # mean
            stat_matrix[fly, 1] = np.std(filter_fly_data) # std
            stat_matrix[fly, 2] = np.median(filter_fly_data) #median
            stat_matrix[fly, 3] = iqr(filter_fly_data) # IQR
            stat_matrix[fly, 4] = len(filter_fly_data) # number of steps

        # store distributions and stats for each genotype
        dist_store.append(fly_dist)
        stat_store.append(stat_matrix)
        
    return dist_store, stat_store
        
        
```

```python
static_stability_BL = []
polygon_area_BL = []
for g in range(len(static_stability_posture)):
    temp_stability = np.copy(static_stability_posture[g])
    temp_area = np.copy(polygon_area_posture[g])
    
    for j in range(static_stability_posture[g].shape[0]):
        temp_stability[j,:] = temp_stability[j,:]/meta_matrix[g][j,-1]
        temp_area[j,:] = temp_area[j,:]/(meta_matrix[g][j,-1]**2)
        
    # append transformed arrays
    static_stability_BL.append(temp_stability)
    polygon_area_BL.append(temp_area)
```

```python
# get distributions and stats for posture parameters
body_height_distribution, body_height_stats = posture_distributions_stats(body_height_posture, meta_matrix, beh_class, -1, False, False)
body_angle_distribution, body_angle_stats = posture_distributions_stats(body_angle, meta_matrix, beh_class, 1, False, True)
body_pitch_distribution, body_pitch_stats = posture_distributions_stats(body_pitch, meta_matrix, beh_class, 1, False, True)
polygon_area_distribution, polygon_area_stats = posture_distributions_stats(polygon_area_BL, meta_matrix, beh_class, 1, True, False)
static_stability_distribution, static_stability_stats = posture_distributions_stats(static_stability_BL, meta_matrix, beh_class, 1, True, False)
```

#### Resting positons of all legs

```python
def resting_bout_measurements(trial_positions, idx1, idx2, n_resting_bouts, trial_polygon):
    n_resting_bouts += 1
        
    # mean x and y positon for each leg
    r1x = np.nanmean(trial_positions[0,0,idx1:idx2])
    r2x = np.nanmean(trial_positions[0,1,idx1:idx2])
    r3x = np.nanmean(trial_positions[0,2,idx1:idx2])
    l1x = np.nanmean(trial_positions[0,3,idx1:idx2])
    l2x = np.nanmean(trial_positions[0,4,idx1:idx2])
    l3x = np.nanmean(trial_positions[0,5,idx1:idx2])
    
    x_resting = np.array([r1x, r2x, r3x, l1x, l2x, l3x])

    r1y = np.nanmean(trial_positions[1,0,idx1:idx2])
    r2y = np.nanmean(trial_positions[1,1,idx1:idx2])
    r3y = np.nanmean(trial_positions[1,2,idx1:idx2])
    l1y = np.nanmean(trial_positions[1,3,idx1:idx2])
    l2y = np.nanmean(trial_positions[1,4,idx1:idx2])
    l3y = np.nanmean(trial_positions[1,5,idx1:idx2])
    
    y_resting = np.array([r1y, r2y, r3y, l1y, l2y, l3y])

    # distance of leg pairs
    front_distance = (np.sqrt(r1x**2 + r1y**2) + np.sqrt(l1x**2 + l1y**2))/2
    mid_distance = (np.sqrt(r2x**2 + r2y**2) + np.sqrt(l2x**2 + l2y**2))/2
    hind_distance = (np.sqrt(r3x**2 + r3y**2) + np.sqrt(l3x**2 + l3y**2))/2
    
    resting_distances = np.array([front_distance, mid_distance, hind_distance])
    
    # mean polygon area
    mean_polygon = np.nanmean(trial_polygon[idx1:idx2])
    
    return n_resting_bouts, x_resting, y_resting, resting_distances, mean_polygon
```

```python
# initialize variables
state = 0
n_resting_bouts = 0

for gen in range(len(genotypes)):
    nflies = len(np.unique(meta_matrix[gen][:, 0]).astype(int).tolist()) # how many flies

    # iterate through flies
    for fly in range(nflies):

        fly_idxs = meta_matrix[gen][:, 0] == fly+1
        sb_idxs = meta_matrix[gen][:, 4] != 1 # just presplit and post split
        trial_idxs = np.logical_and.reduce((fly_idxs, sb_idxs))
        fly_beh_class = beh_class[gen][trial_idxs,:]
        fly_limb_positions = limb_positions[gen][:,:,:,trial_idxs]
        fly_polygon = polygon_area_BL[gen][trial_idxs,:]

        # find resting bouts
        for i in range(fly_beh_class.shape[0]):

            cnt +=1
            idxs = np.where(fly_beh_class[i,:] == 4)[0]
            trial_positions = fly_limb_positions[:,:,:,i]
            trial_polygon = fly_polygon[i,:]
            diff_idxs = np.diff(idxs)
            transitions = np.where(diff_idxs>1)[0]
            if (len(idxs) > 0) and (len(transitions) == 0): # if there is just one long bout
                n_resting_bouts, x_resting, y_resting, resting_distances, mean_polygon = resting_bout_measurements(trial_positions, idxs[0], idxs[-1], n_resting_bouts, trial_polygon)

                if state == 0:
                    state = 1
                    meta_rest = np.array([gen, fly])
                    x_rest_store = x_resting
                    y_rest_store = y_resting
                    distance_rest_store = resting_distances
                    polygon_store = np.array([mean_polygon])
                else:
                    meta_rest = np.vstack((meta_rest, np.array([gen, fly])))
                    x_rest_store = np.vstack((x_rest_store, x_resting))
                    y_rest_store = np.vstack((y_rest_store, y_resting))
                    distance_rest_store = np.vstack((distance_rest_store, resting_distances))
                    polygon_store = np.concatenate((polygon_store, np.array([mean_polygon])))

            if len(transitions) >0:
                if len(transitions) == 1: # just 2 bout:            
                    n_resting_bouts, x_resting, y_resting, resting_distances, mean_polygon = resting_bout_measurements(trial_positions, idxs[0], idxs[transitions[0]], n_resting_bouts, trial_polygon)

                    if state == 0:
                        state = 1
                        meta_rest = np.array([gen, fly])
                        x_rest_store = x_resting
                        y_rest_store = y_resting
                        distance_rest_store = resting_distances
                        polygon_store = np.array([mean_polygon])
                    else:
                        meta_rest = np.vstack((meta_rest, np.array([gen, fly])))
                        x_rest_store = np.vstack((x_rest_store, x_resting))
                        y_rest_store = np.vstack((y_rest_store, y_resting))
                        distance_rest_store = np.vstack((distance_rest_store, resting_distances))
                        polygon_store = np.concatenate((polygon_store, np.array([mean_polygon])))

                    n_resting_bouts, x_resting, y_resting, resting_distances, mean_polygon = resting_bout_measurements(trial_positions, idxs[transitions[0]+1], idxs[-1], n_resting_bouts, trial_polygon)
                    meta_rest = np.vstack((meta_rest, np.array([gen, fly])))
                    x_rest_store = np.vstack((x_rest_store, x_resting))
                    y_rest_store = np.vstack((y_rest_store, y_resting))
                    distance_rest_store = np.vstack((distance_rest_store, resting_distances))
                    polygon_store = np.concatenate((polygon_store, np.array([mean_polygon])))

                else: # multiple bouts
                    start_idx = idxs[0]
                    for j in range(len(transitions)):
                        if len(transitions)-1 == j:
                            end_idx = idxs[transitions[j]]
                            n_resting_bouts, x_resting, y_resting, resting_distances, mean_polygon = resting_bout_measurements(trial_positions, start_idx, end_idx, n_resting_bouts, trial_polygon)

                            if state == 0:
                                state = 1
                                meta_rest = np.array([gen, fly])
                                x_rest_store = x_resting
                                y_rest_store = y_resting
                                distance_rest_store = resting_distances
                                polygon_store = np.array([mean_polygon])
                            else:
                                meta_rest = np.vstack((meta_rest, np.array([gen, fly])))
                                x_rest_store = np.vstack((x_rest_store, x_resting))
                                y_rest_store = np.vstack((y_rest_store, y_resting))
                                distance_rest_store = np.vstack((distance_rest_store, resting_distances))
                                polygon_store = np.concatenate((polygon_store, np.array([mean_polygon])))

                            n_resting_bouts, x_resting, y_resting, resting_distances, mean_polygon = resting_bout_measurements(trial_positions, idxs[transitions[j]+1], idxs[-1], n_resting_bouts, trial_polygon)
                            meta_rest = np.vstack((meta_rest, np.array([gen, fly])))
                            x_rest_store = np.vstack((x_rest_store, x_resting))
                            y_rest_store = np.vstack((y_rest_store, y_resting))
                            distance_rest_store = np.vstack((distance_rest_store, resting_distances))
                            polygon_store = np.concatenate((polygon_store, np.array([mean_polygon])))
                        else:
                            end_idx = idxs[transitions[j]]
                            n_resting_bouts, x_resting, y_resting, resting_distances, mean_polygon = resting_bout_measurements(trial_positions, start_idx, end_idx, n_resting_bouts, trial_polygon)

                            if state == 0:
                                state = 1
                                meta_rest = np.array([gen, fly])
                                x_rest_store = x_resting
                                y_rest_store = y_resting
                                distance_rest_store = resting_distances
                                polygon_store = np.array([mean_polygon])
                            else:
                                meta_rest = np.vstack((meta_rest, np.array([gen, fly])))
                                x_rest_store = np.vstack((x_rest_store, x_resting))
                                y_rest_store = np.vstack((y_rest_store, y_resting))
                                distance_rest_store = np.vstack((distance_rest_store, resting_distances))
                                polygon_store = np.concatenate((polygon_store, np.array([mean_polygon])))

                            start_idx = idxs[transitions[j]+1]

```

#### Static Stability during walking

```python
def walking_posture(metric, meta_matrix, beh_class, sign_clause, zero_filter, neg_ones_filter):
    stat_store = []
    dist_store = []

    # iterate through genotypes
    for gen in range(len(genotypes)):
        nflies = len(np.unique(meta_matrix[gen][:, 0]).astype(int).tolist()) # how many flies

        # create stat matrix
        stat_matrix = np.zeros((nflies, 5))
        fly_dist = []

        # iterate through flies
        for fly in range(nflies):
            #isolate trials of that fly
            fly_idxs = meta_matrix[gen][:, 0] == fly+1
            sb_idxs = meta_matrix[gen][:, 4] != 1 # just presplit and post split
            trial_idxs = np.logical_and.reduce((fly_idxs, sb_idxs))
            fly_data = sign_clause * metric[gen][trial_idxs,:].flatten() # flatten data
            fly_beh_class = beh_class[gen][trial_idxs,:].flatten() 

            # filter on behavior class
            filter_fly_data = fly_data[fly_beh_class==1]
            filter_fly_data = filter_fly_data[~np.isnan(filter_fly_data)] # filter out nans

            # additional filter
            if zero_filter:
                filter_fly_data = filter_fly_data[filter_fly_data != 0] 

            if neg_ones_filter:
                filter_fly_data = filter_fly_data[filter_fly_data != -1]

            # store data of fly-specific metric distrbution and stats
            fly_dist.append(filter_fly_data)
            stat_matrix[fly, 0] = np.mean(filter_fly_data) # mean
            stat_matrix[fly, 1] = np.std(filter_fly_data) # std
            stat_matrix[fly, 2] = np.median(filter_fly_data) #median
            stat_matrix[fly, 3] = iqr(filter_fly_data) # IQR
            stat_matrix[fly, 4] = len(filter_fly_data) # number of steps

        # store distributions and stats for each genotype
        dist_store.append(fly_dist)
        stat_store.append(stat_matrix)
        
    return dist_store, stat_store
```

```python
# get distributions and stats for posture parameters
body_height_distribution_w, body_height_stats_w = walking_posture(body_height_posture, meta_matrix, beh_class, -1, False, False)
polygon_area_distribution_w, polygon_area_stats_w = walking_posture(polygon_area_posture, meta_matrix, beh_class, 1, True, False)
static_stability_distribution_w, static_stability_stats_w = walking_posture(static_stability_posture, meta_matrix, beh_class, 1, True, False)
```

### Intralimb Kinematics - just during forward walking

```python
def intralimb_kinematic_stats_dists(metric, n_step_threshold, meta_matrix, beh_class, tarsi_class):
    # iterate through genotypes
    behavior = 1 # just focus on forward walking
    leg_dists = []
    leg_stats = []
    for j in range(3): # leg combinations
        cnt = 0
        for gen in range(len(genotypes)):
            cnt += 1
            nflies = len(np.unique(meta_matrix[gen][:, 0]).astype(int).tolist()) # how many flies
            tied_idxs = meta_matrix[gen][:, 4] != 1 # presplit + postsplit - combined because no detectable adaptation
            split_idxs = meta_matrix[gen][:, 4] == 1
            metric_stats = np.zeros((nflies, 5, 3, 3)) # fly number, stat value, tied/LS/RS

            # isolate data per leg
            for fly in np.unique(meta_matrix[gen][:, 0]).astype(int).tolist():
                # find all trials associated with the specified conditions
                fly_idxs = meta_matrix[gen][:, 0] == fly # fly

                # find tied, LS, and RS data
                tied_trial_idxs = np.logical_and.reduce((tied_idxs, fly_idxs))
                LS_trial_idxs = np.logical_and.reduce((split_idxs, meta_matrix[gen][:, 3] == 0, fly_idxs))
                RS_trial_idxs = np.logical_and.reduce((split_idxs, meta_matrix[gen][:, 3] == 1, fly_idxs))

                # get data
                # tied - get all presplit + postsplit data
                leg1_tied_data = metric[gen][j, :, tied_trial_idxs].flatten()
                leg2_tied_data = metric[gen][j+3, :, tied_trial_idxs].flatten()

                # combine faster moving data
                leg1_fast_split_data = metric[gen][j, :, RS_trial_idxs].flatten()
                leg2_fast_split_data = metric[gen][j+3, :, LS_trial_idxs].flatten()

                # combine slower moving data
                leg1_slow_split_data = metric[gen][j, :, LS_trial_idxs].flatten() # right legs
                leg2_slow_split_data = metric[gen][j+3, :, RS_trial_idxs].flatten() # left legs

                # behavior classifier
                beh_tied = beh_class[gen][tied_trial_idxs, :].flatten()
                beh_RS = beh_class[gen][RS_trial_idxs, :].flatten()
                beh_LS = beh_class[gen][LS_trial_idxs, :].flatten()

                # tarsi classifier (just for split period)
                tarsi_class_tied = tarsi_class[gen][tied_trial_idxs, :].flatten()
                tarsi_class_RS = tarsi_class[gen][RS_trial_idxs, :].flatten()
                tarsi_class_LS = tarsi_class[gen][LS_trial_idxs, :].flatten()

                # filter tied and split data
                filt_leg1_tied_data = leg1_tied_data[np.logical_and.reduce((beh_tied == behavior, tarsi_class_tied != 7))]
                filt_leg2_tied_data = leg2_tied_data[np.logical_and.reduce((beh_tied == behavior, tarsi_class_tied != 7))]
                cat_tied_data = np.concatenate((filt_leg1_tied_data, filt_leg2_tied_data))

                # filter split data - Fast
                filt_leg1_fast_split_data = leg1_fast_split_data[np.logical_and.reduce((beh_RS == behavior, tarsi_class_RS == 0))]
                filt_leg2_fast_split_data = leg2_fast_split_data[np.logical_and.reduce((beh_LS == behavior, tarsi_class_LS == 0))]
                cat_fast_split_data = np.concatenate((filt_leg1_fast_split_data, filt_leg2_fast_split_data))

                # filter slow split legs
                filt_leg1_slow_split_data = leg1_slow_split_data[np.logical_and.reduce((beh_LS == behavior, tarsi_class_LS == 0))]
                filt_leg2_slow_split_data = leg2_slow_split_data[np.logical_and.reduce((beh_RS == behavior, tarsi_class_RS == 0))]
                cat_slow_split_data = np.concatenate((filt_leg1_slow_split_data, filt_leg2_slow_split_data))

                # filter out nans
                filt_cat_tied_data = cat_tied_data[~np.isnan(cat_tied_data)] 
                filt_cat_fast_split_data = cat_fast_split_data[~np.isnan(cat_fast_split_data)]
                filt_cat_slow_split_data = cat_slow_split_data[~np.isnan(cat_slow_split_data)]

                # filter out -1
                filt_cat_tied_data = filt_cat_tied_data[~(filt_cat_tied_data == -1)]
                filt_cat_fast_split_data = filt_cat_fast_split_data[~(filt_cat_fast_split_data == -1)]
                filt_cat_slow_split_data = filt_cat_slow_split_data[~(filt_cat_slow_split_data == -1)]

                # compute stats
                # tied
                if len(filt_cat_tied_data) < n_step_threshold:
                    metric_stats[fly-1, 0, 1, j] = np.nan
                    metric_stats[fly-1, 1, 1, j] = np.nan
                    metric_stats[fly-1, 2, 1, j] = np.nan
                    metric_stats[fly-1, 3, 1, j] = np.nan
                    metric_stats[fly-1, 4, 1, j] = len(filt_cat_tied_data) # number of steps
                else:
                    metric_stats[fly-1, 0, 1, j] = np.mean(filt_cat_tied_data) # mean
                    metric_stats[fly-1, 1, 1, j] = np.std(filt_cat_tied_data) # std
                    metric_stats[fly-1, 2, 1, j] = np.median(filt_cat_tied_data) #median
                    metric_stats[fly-1, 3, 1, j] = iqr(filt_cat_tied_data) # IQR
                    metric_stats[fly-1, 4, 1, j] = len(filt_cat_tied_data) # number of steps

                # Fast split
                if len(filt_cat_fast_split_data) < n_step_threshold:
                    metric_stats[fly-1, 0, 2, j] = np.nan
                    metric_stats[fly-1, 1, 2, j] = np.nan
                    metric_stats[fly-1, 2, 2, j] = np.nan
                    metric_stats[fly-1, 3, 2, j] = np.nan
                    metric_stats[fly-1, 4, 2, j] = len(filt_cat_fast_split_data) # number of steps
                else:
                    metric_stats[fly-1, 0, 2, j] = np.mean(filt_cat_fast_split_data) # mean
                    metric_stats[fly-1, 1, 2, j] = np.std(filt_cat_fast_split_data) # std
                    metric_stats[fly-1, 2, 2, j] = np.median(filt_cat_fast_split_data) #median
                    metric_stats[fly-1, 3, 2, j] = iqr(filt_cat_fast_split_data) # IQR
                    metric_stats[fly-1, 4, 2, j] = len(filt_cat_fast_split_data) # number of steps

                # Slow split
                if len(filt_cat_slow_split_data) < n_step_threshold:
                    metric_stats[fly-1, 0, 0, j] = np.nan
                    metric_stats[fly-1, 1, 0, j] = np.nan
                    metric_stats[fly-1, 2, 0, j] = np.nan
                    metric_stats[fly-1, 3, 0, j] = np.nan
                    metric_stats[fly-1, 4, 0, j] = len(filt_cat_slow_split_data) # number of steps
                else:
                    metric_stats[fly-1, 0, 0, j] = np.mean(filt_cat_slow_split_data) # mean
                    metric_stats[fly-1, 1, 0, j] = np.std(filt_cat_slow_split_data) # std
                    metric_stats[fly-1, 2, 0, j] = np.median(filt_cat_slow_split_data) #median
                    metric_stats[fly-1, 3, 0, j] = iqr(filt_cat_slow_split_data) # IQR
                    metric_stats[fly-1, 4, 0, j] = len(filt_cat_slow_split_data) # number of steps



                if fly == 1:
                    combined_tied_dist = filt_cat_tied_data 
                    combined_fast_split_dist = filt_cat_fast_split_data
                    combined_slow_split_dist = filt_cat_slow_split_data
                else:
                    combined_tied_dist = np.concatenate((combined_tied_dist, filt_cat_tied_data))
                    combined_fast_split_dist = np.concatenate((combined_fast_split_dist, filt_cat_fast_split_data))
                    combined_slow_split_dist = np.concatenate((combined_slow_split_dist, filt_cat_slow_split_data))

            # store stats
            if j == 0:
                leg_stats.append(metric_stats)
            else:
                leg_stats[gen][:,:,:,j] = metric_stats[:,:,:,j]

            if gen == 0: # first genotype
                genotype_dists = np.concatenate((combined_slow_split_dist, combined_tied_dist, combined_fast_split_dist))
                genotype_order = np.concatenate((np.repeat(cnt, len(combined_slow_split_dist)),np.repeat(cnt+1, len(combined_tied_dist)), np.repeat(cnt+2, len(combined_fast_split_dist))))
                gen_plot_labels = np.concatenate((np.array(['Slow Split - ' + genotype_plot_names[gen] + ' \n(Flies = '+ str(nflies)+', Steps = ' +str(len(combined_slow_split_dist))+')']),
                                                 np.array(['Tied - ' + genotype_plot_names[gen] + ' \n(Flies = '+ str(nflies)+', Steps = ' +str(len(combined_tied_dist))+')']),
                                                 np.array(['Fast Split - ' + genotype_plot_names[gen] + ' \n(Flies = '+ str(nflies)+', Steps = ' +str(len(combined_fast_split_dist))+')'])))
            else:
                genotype_dists = np.concatenate((genotype_dists, np.concatenate((combined_slow_split_dist, combined_tied_dist, combined_fast_split_dist))))
                genotype_order = np.concatenate((genotype_order, np.concatenate((np.repeat(cnt, len(combined_slow_split_dist)), np.repeat(cnt+1, len(combined_tied_dist)), np.repeat(cnt+2, len(combined_fast_split_dist))))))
                temp_labels = np.concatenate((np.array(['Slow Split - ' + genotype_plot_names[gen] + ' \n(Flies = '+ str(nflies)+', Steps = ' +str(len(combined_slow_split_dist))+')']),
                                             np.array(['Tied - ' + genotype_plot_names[gen] + ' \n(Flies = '+ str(nflies)+', Steps = ' +str(len(combined_tied_dist))+')']),
                                             np.array(['Fast Split - ' + genotype_plot_names[gen] + ' \n(Flies = '+ str(nflies)+', Steps = ' +str(len(combined_fast_split_dist))+')'])))
                gen_plot_labels = np.concatenate((gen_plot_labels, temp_labels))
            cnt += 3
        leg_dists.append([genotype_order, genotype_dists, gen_plot_labels])

    return leg_dists, leg_stats
```

```python
def plot_step_kinematics(metric_stats, metric_dist, yname, figure_name, ylims, remove_outliers):

    # distributions
    front_ctrl_dist = metric_dist[0][0] ==2
    front_expt_dist = metric_dist[0][0] ==6
    ctrl_data = metric_dist[0][1][front_ctrl_dist]
    expt_data = metric_dist[0][1][front_expt_dist]
    
    # remove outlier points
    if remove_outliers:
        # control
        outliers = [y for stat in boxplot_stats(ctrl_data) for y in stat['fliers']]
        diff = outliers - np.median(ctrl_data)
        upper_threshold = np.sort(diff[diff>0])[0] + np.median(ctrl_data)   
        ctrl_data = ctrl_data[ctrl_data<upper_threshold]
            
        # expt
        outliers = [y for stat in boxplot_stats(expt_data) for y in stat['fliers']]
        diff = outliers - np.median(expt_data)
        upper_threshold = np.sort(diff[diff>0])[0] + np.median(expt_data)   
        expt_data = expt_data[expt_data<upper_threshold]

    n_steps_ctrl = len(ctrl_data)
    n_steps_expt = len(expt_data)

    # concatenate and give identifiers
    cat_dists = np.concatenate((ctrl_data, expt_data))
    cat_ids = np.concatenate((np.ones(n_steps_ctrl), 2*np.ones(n_steps_expt)))

    # stats
    ctrl_stats = metric_stats[0][:,0,1,0] # mean of each fly
    expt_stats = metric_stats[1][:,0,1,0] # mean of each fly
    cat_means = [ctrl_stats, expt_stats]

    # plot data
    fig = plt.figure(1, figsize = [3,5])

#     sea.violinplot(x=cat_ids, y=cat_dists , order = np.arange(min(cat_ids), max(cat_ids+1)), color = "silver",width = 0.75,
#                   alpha = 0.5, fill=True, inner = None, cut=0) # bw_adjust = 2   

#     # just plot mean
#     lw = 0.2
#     plt.plot(np.array([0-lw + 0+lw]), np.array([np.mean(ctrl_data), np.mean(ctrl_data)]), color = 'black', linewidth = 3.5, zorder = 10)

    # plot individual flies
    for j in range(len(cat_means)):
        x = (j)*np.ones(len(cat_means[j]))+ np.random.choice([-1,1],size=len(cat_means[j]))*np.random.uniform(low=0.0, high=0.10, size=len(cat_means[j]))
        plt.plot(x, cat_means[j], color = 'dimgray', marker= '.', markersize = 10, linestyle = 'none', alpha = 0.75)
        plt.plot(np.array([j-0.2, j+0.2]), np.array([np.mean(cat_dists[cat_ids == j+1]), np.mean(cat_dists[cat_ids == j+1])]),
                color ='black', linewidth = 3.5)

    # reference line for control mean
#     plt.axhline(y=np.mean(cat_dists[cat_ids == 1]), linestyle = 'dotted', linewidth = 1, color = 'black')

    # axes labels
    plt.yticks(fontsize = 12, fontname = 'arial')
    plt.ylabel(yname, fontsize = 14, fontname = 'arial')
    plt.xticks(np.arange(0,2), np.array(['Control (n = '+str(n_steps_ctrl)+')', 'CxHP8 > Kir2.1 (n = '+str(n_steps_expt)+')']), fontsize = 14, fontname = 'Arial', rotation = 45)
    plt.ylim(ylims)
    plt.tight_layout()
    plt.show()

    fig.savefig(figure_dir + figure_name + '.png', dpi = 300)
    fig.savefig(figure_dir + figure_name + '.svg', dpi = 300)
```

### Spatial Plots - AEP, PEP, and stance/swing trajectories

```python
def spatial_kinematics(metric_x, metric_y, meta_matrix, beh_class, tarsi_class, n_step_threshold, filter_metric, lower_filter_threshold, upper_filter_threshold):
    behavior = 1
    leg_dists = []
    leg_stats = []
    for j in range(3): # leg combinations
        cnt = 0
        gen_dist = []
        for gen in range(len(genotypes)):
            cnt += 1
            nflies = len(np.unique(meta_matrix[gen][:, 0]).astype(int).tolist()) # how many flies
            tied_idxs = meta_matrix[gen][:, 4] != 1 # presplit + postsplit - combined because no detectable adaptation
            split_idxs = meta_matrix[gen][:, 4] == 1
            metric_stats = np.zeros((nflies, 5, 3, 3, 2)) # fly number, stat value, tied/LS/RS, leg, x/y

            # isolate data per leg
            for fly in np.unique(meta_matrix[gen][:, 0]).astype(int).tolist():
                # find all trials associated with the specified conditions
                fly_idxs = meta_matrix[gen][:, 0] == fly # fly

                # find tied, LS, and RS data
                tied_trial_idxs = np.logical_and.reduce((tied_idxs, fly_idxs))
                LS_trial_idxs = np.logical_and.reduce((split_idxs, meta_matrix[gen][:, 3] == 0, fly_idxs))
                RS_trial_idxs = np.logical_and.reduce((split_idxs, meta_matrix[gen][:, 3] == 1, fly_idxs))

                '''
                X position
                '''

                # get data
                # tied - get all presplit + postsplit data
                leg1_tied_xdata = metric_x[gen][j, :, tied_trial_idxs].flatten()
                leg2_tied_xdata = metric_x[gen][j+3, :, tied_trial_idxs].flatten()

                # combine faster moving data
                leg1_fast_split_xdata = metric_x[gen][j, :, RS_trial_idxs].flatten()
                leg2_fast_split_xdata = metric_x[gen][j+3, :, LS_trial_idxs].flatten()

                # combine slower moving data
                leg1_slow_split_xdata = metric_x[gen][j, :, LS_trial_idxs].flatten() # right legs
                leg2_slow_split_xdata = metric_x[gen][j+3, :, RS_trial_idxs].flatten() # left legs

                '''
                Y position
                '''
                # get data
                # tied - get all presplit + postsplit data
                leg1_tied_ydata = metric_y[gen][j, :, tied_trial_idxs].flatten()
                leg2_tied_ydata = metric_y[gen][j+3, :, tied_trial_idxs].flatten()

                # combine faster moving data
                leg1_fast_split_ydata = metric_y[gen][j, :, RS_trial_idxs].flatten()
                leg2_fast_split_ydata = metric_y[gen][j+3, :, LS_trial_idxs].flatten()

                # combine slower moving data
                leg1_slow_split_ydata = metric_y[gen][j, :, LS_trial_idxs].flatten() # right legs
                leg2_slow_split_ydata = metric_y[gen][j+3, :, RS_trial_idxs].flatten() # left legs


                '''Filter Metric'''
                leg1_tied_filter_metric = filter_metric[gen][j, :, tied_trial_idxs].flatten()
                leg2_tied_filter_metric = filter_metric[gen][j+3, :, tied_trial_idxs].flatten()

                # combine faster moving data
                leg1_fast_split_filter_metric = filter_metric[gen][j, :, RS_trial_idxs].flatten()
                leg2_fast_split_filter_metric = filter_metric[gen][j+3, :, LS_trial_idxs].flatten()

                # combine slower moving data
                leg1_slow_split_filter_metric = filter_metric[gen][j, :, LS_trial_idxs].flatten() # right legs
                leg2_slow_split_filter_metric = filter_metric[gen][j+3, :, RS_trial_idxs].flatten() # left legs

                # behavior classifier
                beh_tied = beh_class[gen][tied_trial_idxs, :].flatten()
                beh_RS = beh_class[gen][RS_trial_idxs, :].flatten()
                beh_LS = beh_class[gen][LS_trial_idxs, :].flatten()

                # tarsi classifier (just for split period)
                tarsi_class_tied = tarsi_class[gen][tied_trial_idxs, :].flatten()
                tarsi_class_RS = tarsi_class[gen][RS_trial_idxs, :].flatten()
                tarsi_class_LS = tarsi_class[gen][LS_trial_idxs, :].flatten()

                '''Filter Metric'''
                # filter tied and split data
                filt_leg1_tied_filter_metric = leg1_tied_filter_metric[np.logical_and.reduce((beh_tied == behavior, tarsi_class_tied != 7))]
                filt_leg2_tied_filter_metric = leg2_tied_filter_metric[np.logical_and.reduce((beh_tied == behavior, tarsi_class_tied != 7))]
                cat_tied_filter_metric = np.concatenate((filt_leg1_tied_filter_metric, filt_leg2_tied_filter_metric))

                # filter split data - Fast
                filt_leg1_fast_split_filter_metric = leg1_fast_split_filter_metric[np.logical_and.reduce((beh_RS == behavior, tarsi_class_RS == 0))]
                filt_leg2_fast_split_filter_metric = leg2_fast_split_filter_metric[np.logical_and.reduce((beh_LS == behavior, tarsi_class_LS == 0))]
                cat_fast_split_filter_metric = np.concatenate((filt_leg1_fast_split_filter_metric, filt_leg2_fast_split_filter_metric))

                # filter slow split legs
                filt_leg1_slow_split_filter_metric = leg1_slow_split_filter_metric[np.logical_and.reduce((beh_LS == behavior, tarsi_class_LS == 0))]
                filt_leg2_slow_split_filter_metric = leg2_slow_split_filter_metric[np.logical_and.reduce((beh_RS == behavior, tarsi_class_RS == 0))]
                cat_slow_split_filter_metric = np.concatenate((filt_leg1_slow_split_filter_metric, filt_leg2_slow_split_filter_metric))

                # filter tied and split data
                filt_leg1_tied_xdata = leg1_tied_xdata[np.logical_and.reduce((beh_tied == behavior, tarsi_class_tied != 7))]
                filt_leg2_tied_xdata = leg2_tied_xdata[np.logical_and.reduce((beh_tied == behavior, tarsi_class_tied != 7))]
                filt_leg1_tied_ydata = leg1_tied_ydata[np.logical_and.reduce((beh_tied == behavior, tarsi_class_tied != 7))]
                filt_leg2_tied_ydata = leg2_tied_ydata[np.logical_and.reduce((beh_tied == behavior, tarsi_class_tied != 7))]
                cat_tied_xdata = np.concatenate((filt_leg1_tied_xdata, filt_leg2_tied_xdata))
                cat_tied_ydata = np.concatenate((-filt_leg1_tied_ydata, filt_leg2_tied_ydata))

                # filter split data - Fast
                filt_leg1_fast_split_xdata = leg1_fast_split_xdata[np.logical_and.reduce((beh_RS == behavior, tarsi_class_RS == 0))]
                filt_leg2_fast_split_xdata = leg2_fast_split_xdata[np.logical_and.reduce((beh_LS == behavior, tarsi_class_LS == 0))]
                filt_leg1_fast_split_ydata = leg1_fast_split_ydata[np.logical_and.reduce((beh_RS == behavior, tarsi_class_RS == 0))]
                filt_leg2_fast_split_ydata = leg2_fast_split_ydata[np.logical_and.reduce((beh_LS == behavior, tarsi_class_LS == 0))]
                cat_fast_split_xdata = np.concatenate((filt_leg1_fast_split_xdata, filt_leg2_fast_split_xdata))
                cat_fast_split_ydata = np.concatenate((-filt_leg1_fast_split_ydata, filt_leg2_fast_split_ydata))

                # filter slow split legs
                filt_leg1_slow_split_xdata = leg1_slow_split_xdata[np.logical_and.reduce((beh_LS == behavior, tarsi_class_LS == 0))]
                filt_leg2_slow_split_xdata = leg2_slow_split_xdata[np.logical_and.reduce((beh_RS == behavior, tarsi_class_RS == 0))]
                filt_leg1_slow_split_ydata = leg1_slow_split_ydata[np.logical_and.reduce((beh_LS == behavior, tarsi_class_LS == 0))]
                filt_leg2_slow_split_ydata = leg2_slow_split_ydata[np.logical_and.reduce((beh_RS == behavior, tarsi_class_RS == 0))]
                cat_slow_split_xdata = np.concatenate((filt_leg1_slow_split_xdata, filt_leg2_slow_split_xdata))
                cat_slow_split_ydata = np.concatenate((-filt_leg1_slow_split_ydata, filt_leg2_slow_split_ydata))

                '''Filter data based on filter metric'''
                cat_tied_xdata[np.logical_or(cat_tied_filter_metric<lower_filter_threshold, cat_tied_filter_metric>upper_filter_threshold)]=-5
                cat_fast_split_xdata[np.logical_or(cat_fast_split_filter_metric<lower_filter_threshold, cat_fast_split_filter_metric>upper_filter_threshold)]=-5
                cat_slow_split_xdata[np.logical_or(cat_slow_split_filter_metric<lower_filter_threshold, cat_slow_split_filter_metric>upper_filter_threshold)]=-5

                cat_tied_ydata[np.logical_or(cat_tied_filter_metric<lower_filter_threshold, cat_tied_filter_metric>upper_filter_threshold)]=-5
                cat_fast_split_ydata[np.logical_or(cat_fast_split_filter_metric<lower_filter_threshold, cat_fast_split_filter_metric>upper_filter_threshold)]=-5
                cat_slow_split_ydata[np.logical_or(cat_slow_split_filter_metric<lower_filter_threshold, cat_slow_split_filter_metric>upper_filter_threshold)]=-5

                # filter out nans
                filt_cat_tied_xdata = cat_tied_xdata[~np.isnan(cat_tied_xdata)] 
                filt_cat_fast_split_xdata = cat_fast_split_xdata[~np.isnan(cat_fast_split_xdata)]
                filt_cat_slow_split_xdata = cat_slow_split_xdata[~np.isnan(cat_slow_split_xdata)]

                filt_cat_tied_ydata = cat_tied_ydata[~np.isnan(cat_tied_ydata)] 
                filt_cat_fast_split_ydata = cat_fast_split_ydata[~np.isnan(cat_fast_split_ydata)]
                filt_cat_slow_split_ydata = cat_slow_split_ydata[~np.isnan(cat_slow_split_ydata)]

                # filter out the metric value
                # filter out -5
                filt_cat_tied_xdata = filt_cat_tied_xdata[~(filt_cat_tied_xdata == -5)]
                filt_cat_fast_split_xdata = filt_cat_fast_split_xdata[~(filt_cat_fast_split_xdata == -5)]
                filt_cat_slow_split_xdata = filt_cat_slow_split_xdata[~(filt_cat_slow_split_xdata == -5)]

                filt_cat_tied_ydata = filt_cat_tied_ydata[~(filt_cat_tied_ydata == -5)]
                filt_cat_fast_split_ydata = filt_cat_fast_split_ydata[~(filt_cat_fast_split_ydata == -5)]
                filt_cat_slow_split_ydata = filt_cat_slow_split_ydata[~(filt_cat_slow_split_ydata == -5)]

                # compute stats
                # tied
                if len(filt_cat_tied_xdata) <= n_step_threshold:
                    metric_stats[fly-1, 0, 1, j, 0] = np.nan
                    metric_stats[fly-1, 1, 1, j, 0] = np.nan
                    metric_stats[fly-1, 2, 1, j, 0] = np.nan
                    metric_stats[fly-1, 3, 1, j, 0] = np.nan
                    metric_stats[fly-1, 4, 1, j, 0] = len(filt_cat_tied_xdata) # number of steps

                    metric_stats[fly-1, 0, 1, j, 1] = np.nan
                    metric_stats[fly-1, 1, 1, j, 1] = np.nan
                    metric_stats[fly-1, 2, 1, j, 1] = np.nan
                    metric_stats[fly-1, 3, 1, j, 1] = np.nan
                    metric_stats[fly-1, 4, 1, j, 1] = len(filt_cat_tied_ydata) # number of steps
                else:
                    metric_stats[fly-1, 0, 1, j, 0] = np.mean(filt_cat_tied_xdata) # mean
                    metric_stats[fly-1, 1, 1, j, 0] = np.std(filt_cat_tied_xdata) # std
                    metric_stats[fly-1, 2, 1, j, 0] = np.median(filt_cat_tied_xdata) #median
                    metric_stats[fly-1, 3, 1, j, 0] = iqr(filt_cat_tied_xdata) # IQR
                    metric_stats[fly-1, 4, 1, j, 0] = len(filt_cat_tied_xdata) # number of steps

                    metric_stats[fly-1, 0, 1, j, 1] = np.mean(filt_cat_tied_ydata) # mean
                    metric_stats[fly-1, 1, 1, j, 1] = np.std(filt_cat_tied_ydata) # std
                    metric_stats[fly-1, 2, 1, j, 1] = np.median(filt_cat_tied_ydata) #median
                    metric_stats[fly-1, 3, 1, j, 1] = iqr(filt_cat_tied_ydata) # IQR
                    metric_stats[fly-1, 4, 1, j, 1] = len(filt_cat_tied_ydata) # number of steps

                # Fast split
                if len(filt_cat_fast_split_xdata) <= n_step_threshold:
                    metric_stats[fly-1, 0, 2, j,0] = np.nan
                    metric_stats[fly-1, 1, 2, j,0] = np.nan
                    metric_stats[fly-1, 2, 2, j,0] = np.nan
                    metric_stats[fly-1, 3, 2, j,0] = np.nan
                    metric_stats[fly-1, 4, 2, j,0] = len(filt_cat_fast_split_xdata) # number of steps

                    metric_stats[fly-1, 0, 2, j,1] = np.nan
                    metric_stats[fly-1, 1, 2, j,1] = np.nan
                    metric_stats[fly-1, 2, 2, j,1] = np.nan
                    metric_stats[fly-1, 3, 2, j,1] = np.nan
                    metric_stats[fly-1, 4, 2, j,1] = len(filt_cat_fast_split_ydata) # number of steps
                else:
                    metric_stats[fly-1, 0, 2, j,0] = np.mean(filt_cat_fast_split_xdata) # mean
                    metric_stats[fly-1, 1, 2, j,0] = np.std(filt_cat_fast_split_xdata) # std
                    metric_stats[fly-1, 2, 2, j,0] = np.median(filt_cat_fast_split_xdata) #median
                    metric_stats[fly-1, 3, 2, j,0] = iqr(filt_cat_fast_split_xdata) # IQR
                    metric_stats[fly-1, 4, 2, j,0] = len(filt_cat_fast_split_xdata) # number of steps

                    metric_stats[fly-1, 0, 2, j,1] = np.mean(filt_cat_fast_split_ydata) # mean
                    metric_stats[fly-1, 1, 2, j,1] = np.std(filt_cat_fast_split_ydata) # std
                    metric_stats[fly-1, 2, 2, j,1] = np.median(filt_cat_fast_split_ydata) #median
                    metric_stats[fly-1, 3, 2, j,1] = iqr(filt_cat_fast_split_ydata) # IQR
                    metric_stats[fly-1, 4, 2, j,1] = len(filt_cat_fast_split_ydata) # number of steps

                # Slow split
                if len(filt_cat_slow_split_xdata) <= n_step_threshold:
                    metric_stats[fly-1, 0, 0, j,0] = np.nan
                    metric_stats[fly-1, 1, 0, j,0] = np.nan
                    metric_stats[fly-1, 2, 0, j,0] = np.nan
                    metric_stats[fly-1, 3, 0, j,0] = np.nan
                    metric_stats[fly-1, 4, 0, j,0] = len(filt_cat_slow_split_xdata) # number of steps

                    metric_stats[fly-1, 0, 0, j,1] = np.nan
                    metric_stats[fly-1, 1, 0, j,1] = np.nan
                    metric_stats[fly-1, 2, 0, j,1] = np.nan
                    metric_stats[fly-1, 3, 0, j,1] = np.nan
                    metric_stats[fly-1, 4, 0, j,1] = len(filt_cat_slow_split_ydata) # number of steps
                else:
                    metric_stats[fly-1, 0, 0, j,0] = np.mean(filt_cat_slow_split_xdata) # mean
                    metric_stats[fly-1, 1, 0, j,0] = np.std(filt_cat_slow_split_xdata) # std
                    metric_stats[fly-1, 2, 0, j,0] = np.median(filt_cat_slow_split_xdata) #median
                    metric_stats[fly-1, 3, 0, j,0] = iqr(filt_cat_slow_split_xdata) # IQR
                    metric_stats[fly-1, 4, 0, j,0] = len(filt_cat_slow_split_xdata) # number of steps

                    metric_stats[fly-1, 0, 0, j,1] = np.mean(filt_cat_slow_split_ydata) # mean
                    metric_stats[fly-1, 1, 0, j,1] = np.std(filt_cat_slow_split_ydata) # std
                    metric_stats[fly-1, 2, 0, j,1] = np.median(filt_cat_slow_split_ydata) #median
                    metric_stats[fly-1, 3, 0, j,1] = iqr(filt_cat_slow_split_ydata) # IQR
                    metric_stats[fly-1, 4, 0, j,1] = len(filt_cat_slow_split_ydata) # number of steps


                if fly == 1:
                    combined_tied_distx = filt_cat_tied_xdata 
                    combined_fast_split_distx = filt_cat_fast_split_xdata
                    combined_slow_split_distx = filt_cat_slow_split_xdata

                    combined_tied_disty = filt_cat_tied_ydata 
                    combined_fast_split_disty = filt_cat_fast_split_ydata
                    combined_slow_split_disty = filt_cat_slow_split_ydata
                else:
                    combined_tied_distx = np.concatenate((combined_tied_distx, filt_cat_tied_xdata))
                    combined_fast_split_distx = np.concatenate((combined_fast_split_distx, filt_cat_fast_split_xdata))
                    combined_slow_split_distx = np.concatenate((combined_slow_split_distx, filt_cat_slow_split_xdata))

                    combined_tied_disty = np.concatenate((combined_tied_disty, filt_cat_tied_ydata))
                    combined_fast_split_disty = np.concatenate((combined_fast_split_disty, filt_cat_fast_split_ydata))
                    combined_slow_split_disty = np.concatenate((combined_slow_split_disty, filt_cat_slow_split_ydata))

            # store stats
            if j == 0:
                leg_stats.append(metric_stats)
            else:
                leg_stats[gen][:,:,:,j,0] = metric_stats[:,:,:,j,0]
                leg_stats[gen][:,:,:,j,1] = metric_stats[:,:,:,j,1]

            gen_dist.append([np.vstack((combined_slow_split_distx, combined_slow_split_disty)), np.vstack((combined_tied_distx, combined_tied_disty)), np.vstack((combined_fast_split_distx, combined_fast_split_disty))])

        leg_dists.append(gen_dist)
       
    return leg_dists, leg_stats
```

### Stance and Swing Trajectories

```python
# find the transition frames of a walking bout
def find_walking_bouts(beh_frames):
    start_frame = np.array([beh_frames[0]])
    for frame in range(len(beh_frames)-1):
        if (beh_frames[frame+1] - beh_frames[frame]) > 1:
            if len(start_frame) == 1:
                end_frame = np.array([beh_frames[frame]])
                start_frame = np.concatenate((start_frame, np.array([beh_frames[frame+1]])))
            else:
                start_frame = np.concatenate((start_frame, np.array([beh_frames[frame+1]])))
                end_frame = np.concatenate((end_frame,np.array([beh_frames[frame]])))
    if ((beh_frames[frame+1] - beh_frames[frame]) == 1) and len(start_frame) == 1:
        end_frame = np.array([beh_frames[-1]])
    else:
        end_frame = np.concatenate((end_frame,np.array([beh_frames[-1]])))
    walking_bout_transitions = np.vstack((start_frame, end_frame))
    
    return walking_bout_transitions                                    
```

```python
def stance_swing_bouts(bout_swing_stance):
    stance_starts = np.where(np.diff(bout_swing_stance) ==1)[0]+1 #swing to stance
    if bout_swing_stance[0] != 1:
        stance_ends = np.where(np.diff(bout_swing_stance) ==-1)[0]
    else:
        try:
            stance_ends = np.where(np.diff(bout_swing_stance) ==-1)[0][1::] # ignore the first end (edge effect)
        except:
            stance_ends = np.array([])

    # find the corresponding stance start to the stance ends
    threshold = 3 # need 5 or more samples ~ 25 ms
    max_threshold = 30 # need 5 or more samples ~ 25 ms
    if len(stance_ends)>0: # make sure there is a stance bout
        temp_stance_bouts = []
        for j in range(len(stance_ends)):
            diff_starts = stance_ends[j] - stance_starts
            if len(diff_starts[np.logical_and(diff_starts>threshold+1, diff_starts<=max_threshold)]) > 0: # place an upper bound
                min_diff_idx = np.argmin(diff_starts[diff_starts>threshold])
                resize_stances = stance_starts[diff_starts>threshold]
                stance_start_match = resize_stances[min_diff_idx]
                temp_stance_bouts.append([stance_start_match, stance_ends[j]])
        # reformat stance bouts
        if len(temp_stance_bouts)>0:
            stance_bouts = np.zeros((2, len(temp_stance_bouts)))
            for k in range(len(temp_stance_bouts)):
                stance_bouts[0,k] = temp_stance_bouts[k][0]
                stance_bouts[1,k] = temp_stance_bouts[k][1]
        else:
            stance_bouts = np.array([]) 
    else:
        stance_bouts = np.array([])

    # REPEAT FOR SWING
    swing_starts = np.where(np.diff(bout_swing_stance) ==-1)[0]+1 #swing to stance
    if bout_swing_stance[0] != 0:
        swing_ends = np.where(np.diff(bout_swing_stance) ==1)[0]
    else:
        try:
            swing_ends = np.where(np.diff(bout_swing_stance) ==1)[0][1::] # ignore the first end (edge effect)
        except:
            swing_ends = np.array([])
    
    swing_max_threshold = 15 # 75 ms
    if len(swing_ends)>0: # make sure there is a stance bout
        temp_swing_bouts = []
        for j in range(len(swing_ends)):
            diff_starts = swing_ends[j] - swing_starts
            if len(diff_starts[np.logical_and(diff_starts>threshold, diff_starts<=swing_max_threshold)]) > 0: # need ~20ms
                min_diff_idx = np.argmin(diff_starts[diff_starts>threshold])
                resize_swing = swing_starts[diff_starts>threshold]
                swing_start_match = resize_swing[min_diff_idx]
                temp_swing_bouts.append([swing_start_match, swing_ends[j]])
        # reformat stance bouts
        if len(temp_swing_bouts)>0:
            swing_bouts = np.zeros((2, len(temp_swing_bouts)))
            for k in range(len(temp_swing_bouts)):
                swing_bouts[0,k] = temp_swing_bouts[k][0]
                swing_bouts[1,k] = temp_swing_bouts[k][1]
        else:
            swing_bouts = np.array([]) 
    else:
        swing_bouts = np.array([])
            
    return stance_bouts, swing_bouts
```

```python
behavior_trajs = []
behavior_min_max = []
walking_bout_threshold = 40 # walking about more 200 ms 
for behavior in [1]: # forward walking, Left Turn, Right Turn
    gen_trajs = []
    min_max_l_array = np.zeros((6,2,len(limb_positions),2)) # legs, min/max, genotype, stance/swing
    for gen in range(len(limb_positions)):
        
        nflies = len(np.unique(meta_matrix[gen][:, 0]).astype(int).tolist()) # how many flies
        tied_idxs = meta_matrix[gen][:, 4] != 1 # presplit + postsplit - combined because no detectable adaptation
        split_idxs = meta_matrix[gen][:, 4] == 1
        
        leg_stance_trajs = []
        leg_swing_trajs = []
        for leg in range(6):
            fly_stance_trajs_tied = []
            fly_swing_trajs_tied = []
            fly_stance_trajs_LS = []
            fly_swing_trajs_LS = []
            fly_stance_trajs_RS = []
            fly_swing_trajs_RS = []
            
            min_max_stance = 0 # find the min and max for each leg
            min_max_swing = 0 # find the min and max for each leg
            
            for fly in np.unique(meta_matrix[gen][:, 0]).astype(int).tolist():
                stance_trajs_tied = []
                swing_trajs_tied = []
                stance_trajs_LS = []
                swing_trajs_LS = []
                stance_trajs_RS = []
                swing_trajs_RS = []
                
                # find all trials associated with the specified conditions
                fly_idxs = meta_matrix[gen][:, 0] == fly # fly

                # find tied, LS, and RS data
                tied_trial_idxs = np.logical_and.reduce((tied_idxs, fly_idxs))
                LS_trial_idxs = np.logical_and.reduce((split_idxs, meta_matrix[gen][:, 3] == 0, fly_idxs))
                RS_trial_idxs = np.logical_and.reduce((split_idxs, meta_matrix[gen][:, 3] == 1, fly_idxs))
                
                # isolate data for each of the split conditions
                tied_x = limb_positions[gen][0][leg,:,tied_trial_idxs]
                tied_y = limb_positions[gen][1][leg,:,tied_trial_idxs]
                tied_z = limb_positions[gen][2][leg,:,tied_trial_idxs]
                
                LS_x = limb_positions[gen][0][leg,:,LS_trial_idxs]
                LS_y = limb_positions[gen][1][leg,:,LS_trial_idxs]
                LS_z = limb_positions[gen][2][leg,:,LS_trial_idxs]
                
                RS_x = limb_positions[gen][0][leg,:,RS_trial_idxs]
                RS_y = limb_positions[gen][1][leg,:,RS_trial_idxs]
                RS_z = limb_positions[gen][2][leg,:,RS_trial_idxs]
                
                # swing stance matrix
                swing_stance_tied = swing_stance_mat[gen][leg,:,tied_trial_idxs]
                swing_stance_LS = swing_stance_mat[gen][leg,:,LS_trial_idxs]
                swing_stance_RS = swing_stance_mat[gen][leg,:,RS_trial_idxs]
                
                # behavior classifier
                beh_tied = beh_class[gen][tied_trial_idxs, :]
                beh_RS = beh_class[gen][RS_trial_idxs, :]
                beh_LS = beh_class[gen][LS_trial_idxs, :]

                # tarsi classifier (just for split period)
                tarsi_class_tied = tarsi_class[gen][tied_trial_idxs, :]
                tarsi_class_RS = tarsi_class[gen][RS_trial_idxs, :]
                tarsi_class_LS = tarsi_class[gen][LS_trial_idxs, :]
                
                # each trail should be treated independently (no false continuation of a step)
                '''Tied'''
                for i in range(swing_stance_tied.shape[0]):
                    trial_swing_stance = swing_stance_tied[i,:]
                    
                    # find frames where behavior is observed
                    beh_frames = np.where(np.logical_and.reduce((beh_tied[i,:] == behavior, tarsi_class_tied[i,:] != 7))==True)[0]
                    
                    # walking bouts
                    if len(beh_frames) > walking_bout_threshold : # make sure there are respective walking frames
                        walking_bout_transitions = find_walking_bouts(beh_frames)
                    
                        # only anlyze stance/swing trajectories in the middle of the walking bout (middle steps)
                        for bout in range(walking_bout_transitions.shape[1]):
                            bout_swing_stance = trial_swing_stance[walking_bout_transitions[0,bout]: walking_bout_transitions[1,bout]+1]

                            # within the bout find the inner stances (1) or swings (0)
                            stance_bouts, swing_bouts = stance_swing_bouts(bout_swing_stance)

                            # if there are stance and swing bouts, isolate the corresponding trajectories
                            # stance trajectories - just x and y (no change in z)
                            if len(stance_bouts)>0:
                                for j in range(stance_bouts.shape[1]):
                                    stance_start_frame = np.arange(walking_bout_transitions[0,bout], walking_bout_transitions[1,bout]+1)[int(stance_bouts[0,j])]
                                    stance_end_frame = np.arange(walking_bout_transitions[0,bout], walking_bout_transitions[1,bout]+1)[int(stance_bouts[1,j])+1] # adjust for end
                                    curr_x = tied_x[i, stance_start_frame:stance_end_frame]
                                    curr_y = tied_y[i, stance_start_frame:stance_end_frame]

                                    # find min and max values during stance
                                    stance_trajs_tied.append([curr_x, curr_y])
                                    if min_max_stance == 0:
                                        min_stance_x = np.min(curr_x)
                                        max_stance_x = np.max(curr_x)
                                        min_max_stance = 1
                                    else:
                                        if np.min(curr_x) < min_stance_x:
                                            min_stance_x = np.min(curr_x)
                                        if np.max(curr_x) > max_stance_x:
                                            max_stance_x = np.max(curr_x)

                                
                            # swing trajectories - just x and y (no change in z)
                            if len(swing_bouts)>0:
                                for j in range(swing_bouts.shape[1]):
                                    swing_start_frame = np.arange(walking_bout_transitions[0,bout], walking_bout_transitions[1,bout]+1)[int(swing_bouts[0,j])]-1
                                    swing_end_frame = np.arange(walking_bout_transitions[0,bout], walking_bout_transitions[1,bout]+1)[int(swing_bouts[1,j])+1]+1 # adjust for end
                                    curr_x = tied_x[i, swing_start_frame:swing_end_frame]
                                    curr_y = tied_y[i, swing_start_frame:swing_end_frame]
                                    curr_z = tied_z[i, swing_start_frame:swing_end_frame]

                                    # find min and max values during stance
                                    swing_trajs_tied.append([curr_x, curr_y, curr_z])
                                    if min_max_swing == 0:
                                        min_swing_x = np.min(curr_x)
                                        max_swing_x = np.max(curr_x)
                                        min_max_swing = 1
                                    else:
                                        if np.min(curr_x) < min_swing_x:
                                            min_swing_x = np.min(curr_x)
                                        if np.max(curr_x) > max_swing_x:
                                            max_swing_x = np.max(curr_x)

                '''LS'''
                for i in range(swing_stance_LS.shape[0]):
                    trial_swing_stance = swing_stance_LS[i,:]
                    
                    # find frames where behavior is observed
                    if behavior == 1: # during forward walking - consider only ture split
                        beh_frames = np.where(np.logical_and.reduce((beh_LS[i,:] == behavior, tarsi_class_LS[i,:] == 0))==True)[0]
                    else:
                        beh_frames = np.where(np.logical_and.reduce((beh_LS[i,:] == behavior, tarsi_class_LS[i,:] != 7))==True)[0]
                    
                    # walking bouts
                    if len(beh_frames) > walking_bout_threshold : # make sure there are respective walking frames
                        walking_bout_transitions = find_walking_bouts(beh_frames)
                    
                        # only anlyze stance/swing trajectories in the middle of the walking bout (middle steps)
                        for bout in range(walking_bout_transitions.shape[1]):
                            bout_swing_stance = trial_swing_stance[walking_bout_transitions[0,bout]: walking_bout_transitions[1,bout]+1]

                            # within the bout find the inner stances (1) or swings (0)
                            stance_bouts, swing_bouts = stance_swing_bouts(bout_swing_stance)

                            # if there are stance and swing bouts, isolate the corresponding trajectories
                            # stance trajectories - just x and y (no change in z)
                            if len(stance_bouts)>0:
                                for j in range(stance_bouts.shape[1]):
                                    stance_start_frame = np.arange(walking_bout_transitions[0,bout], walking_bout_transitions[1,bout]+1)[int(stance_bouts[0,j])]
                                    stance_end_frame = np.arange(walking_bout_transitions[0,bout], walking_bout_transitions[1,bout]+1)[int(stance_bouts[1,j])+1] # adjust for end
                                    curr_x = LS_x[i, stance_start_frame:stance_end_frame]
                                    curr_y = LS_y[i, stance_start_frame:stance_end_frame]

                                    # find min and max values during stance
                                    stance_trajs_LS.append([curr_x, curr_y])
                                    if min_max_stance == 0:
                                        min_stance_x = np.min(curr_x)
                                        max_stance_x = np.max(curr_x)
                                        min_max_stance = 1
                                    else:
                                        if np.min(curr_x) < min_stance_x:
                                            min_stance_x = np.min(curr_x)
                                        if np.max(curr_x) > max_stance_x:
                                            max_stance_x = np.max(curr_x)


                            # swing trajectories - just x and y (no change in z)
                            if len(swing_bouts)>0:
                                for j in range(swing_bouts.shape[1]):
                                    swing_start_frame = np.arange(walking_bout_transitions[0,bout], walking_bout_transitions[1,bout]+1)[int(swing_bouts[0,j])]-1
                                    swing_end_frame = np.arange(walking_bout_transitions[0,bout], walking_bout_transitions[1,bout]+1)[int(swing_bouts[1,j])+1]+1 # adjust for end
                                    curr_x = LS_x[i, swing_start_frame:swing_end_frame]
                                    curr_y = LS_y[i, swing_start_frame:swing_end_frame]
                                    curr_z = LS_z[i, swing_start_frame:swing_end_frame]

                                    # find min and max values during stance
                                    swing_trajs_LS.append([curr_x, curr_y, curr_z])
                                    if min_max_swing == 0:
                                        min_swing_x = np.min(curr_x)
                                        max_swing_x = np.max(curr_x)
                                        min_max_swing = 1
                                    else:
                                        if np.min(curr_x) < min_swing_x:
                                            min_swing_x = np.min(curr_x)
                                        if np.max(curr_x) > max_swing_x:
                                            max_swing_x = np.max(curr_x)
                                        
                '''RS'''
                for i in range(swing_stance_RS.shape[0]):
                    trial_swing_stance = swing_stance_RS[i,:]
                    
                    # find frames where behavior is observed
                    if behavior == 1: # during forward walking - consider only ture split
                        beh_frames = np.where(np.logical_and.reduce((beh_RS[i,:] == behavior, tarsi_class_RS[i,:] == 0))==True)[0]
                    else:
                        beh_frames = np.where(np.logical_and.reduce((beh_RS[i,:] == behavior, tarsi_class_RS[i,:] != 7))==True)[0]
                    
                    # walking bouts
                    if len(beh_frames) > walking_bout_threshold : # make sure there are respective walking frames
                        walking_bout_transitions = find_walking_bouts(beh_frames)
                    
                        # only anlyze stance/swing trajectories in the middle of the walking bout (middle steps)
                        for bout in range(walking_bout_transitions.shape[1]):
                            bout_swing_stance = trial_swing_stance[walking_bout_transitions[0,bout]: walking_bout_transitions[1,bout]+1]

                            # within the bout find the inner stances (1) or swings (0)
                            stance_bouts, swing_bouts = stance_swing_bouts(bout_swing_stance)

                            # if there are stance and swing bouts, isolate the corresponding trajectories
                            # stance trajectories - just x and y (no change in z)
                            if len(stance_bouts)>0:
                                for j in range(stance_bouts.shape[1]):
                                    stance_start_frame = np.arange(walking_bout_transitions[0,bout], walking_bout_transitions[1,bout]+1)[int(stance_bouts[0,j])]
                                    stance_end_frame = np.arange(walking_bout_transitions[0,bout], walking_bout_transitions[1,bout]+1)[int(stance_bouts[1,j])+1] # adjust for end
                                    curr_x = RS_x[i, stance_start_frame:stance_end_frame]
                                    curr_y = RS_y[i, stance_start_frame:stance_end_frame]

                                    # find min and max values during stance
                                    stance_trajs_RS.append([curr_x, curr_y])
                                    if min_max_stance == 0:
                                        min_stance_x = np.min(curr_x)
                                        max_stance_x = np.max(curr_x)
                                        min_max_stance = 1
                                    else:
                                        if np.min(curr_x) < min_stance_x:
                                            min_stance_x = np.min(curr_x)
                                        if np.max(curr_x) > max_stance_x:
                                            max_stance_x = np.max(curr_x)


                            # swing trajectories - just x and y (no change in z)
                            if len(swing_bouts)>0:
                                for j in range(swing_bouts.shape[1]):
                                    swing_start_frame = np.arange(walking_bout_transitions[0,bout], walking_bout_transitions[1,bout]+1)[int(swing_bouts[0,j])]-1
                                    swing_end_frame = np.arange(walking_bout_transitions[0,bout], walking_bout_transitions[1,bout]+1)[int(swing_bouts[1,j])+1]+1 # adjust for end
                                    curr_x = RS_x[i, swing_start_frame:swing_end_frame]
                                    curr_y = RS_y[i, swing_start_frame:swing_end_frame]
                                    curr_z = RS_z[i, swing_start_frame:swing_end_frame]

                                    # find min and max values during stance
                                    swing_trajs_RS.append([curr_x, curr_y, curr_z])
                                    if min_max_swing == 0:
                                        min_swing_x = np.min(curr_x)
                                        max_swing_x = np.max(curr_x)
                                        min_max_swing = 1
                                    else:
                                        if np.min(curr_x) < min_swing_x:
                                            min_swing_x = np.min(curr_x)
                                        if np.max(curr_x) > max_swing_x:
                                            max_swing_x = np.max(curr_x)

                # append trajectories for each fly
                fly_stance_trajs_tied.append(stance_trajs_tied)
                fly_swing_trajs_tied.append(swing_trajs_tied)
                fly_stance_trajs_LS.append(stance_trajs_LS)
                fly_swing_trajs_LS.append(swing_trajs_LS)
                fly_stance_trajs_RS.append(stance_trajs_RS)
                fly_swing_trajs_RS.append(swing_trajs_RS)
                
            # append trajectories for each leg
            leg_stance_trajs.append([fly_stance_trajs_LS, fly_stance_trajs_tied, fly_stance_trajs_RS])
            leg_swing_trajs.append([fly_swing_trajs_LS, fly_swing_trajs_tied, fly_swing_trajs_RS])
            
            # track minimum an maximum x values
            min_max_l_array[leg,0,gen,0] = min_stance_x 
            min_max_l_array[leg,1,gen,0] = max_stance_x 
            min_max_l_array[leg,0,gen,1] = min_swing_x 
            min_max_l_array[leg,1,gen,1] = max_swing_x 
            
            
        # append data for genotypes
        gen_trajs.append([leg_stance_trajs, leg_swing_trajs])
        
    # save behavior trajectories (Forward walking and left/right turning)
    behavior_trajs.append(gen_trajs)
    behavior_min_max.append(min_max_l_array)
```

```python
# calculate mean of polar distributions by transforming them into cartesian coordinates then back to polar coordinates
def estimate_ang_stats(var):
#     mean_angle = np.arctan2(np.nanmedian(np.sin(var)), np.nanmedian(np.cos(var)))
    mean_angle = np.arctan2(np.nanmean(np.sin(var)), np.nanmean(np.cos(var)))
    angular_dispersion = np.sqrt(np.nanmean(np.sin(var))**2 + np.nanmean(np.cos(var))**2) # 0: uniform, 1: concetrated in one directtion
    return mean_angle, angular_dispersion
```

```python
def compute_histogram(var, N):
    hist, bin_edges = np.histogram(var, bins=N, range=(0.0,2*np.pi), density=True)
    bin_centers = bin_edges[1::] - ((bin_edges[1]- bin_edges[0])/2)
    return hist, bin_centers 
```

```python
# transform phase values to between -np.pi to np.pi
def transform_half_polar(var):
    var[var>np.pi] = var[var>np.pi] - (2*np.pi)
    return var
```

```python
def full_distribution_transform_polar(var):
    var[var<0] = np.pi + np.abs(-np.pi - var[var<0])
    return var
```

```python
def compute_inphase_histogram(var, N):
    hist, bin_edges = np.histogram(var, bins=N, range=(-np.pi, np.pi), density=True)
    bin_centers = bin_edges[1::] - ((bin_edges[1]- bin_edges[0])/2)
    return hist, bin_centers 
```

```python
# function that transforms that space from 0 to pi to 0 to 2pi
def transform_polar(org_val):
    if org_val < 0: # only need to transform negative values to extend the space to 2pi
        transformed_val = np.pi + np.abs(-np.pi - org_val)
    else: 
        transformed_val = org_val
        
    return transformed_val
```

```python
# compute phase stats and distributions
step_threshold = 10 # empirically determined
Nbins = 15
rel_phase_comparasions = [1, 4, 0, 3, 2] # L1-R2, L1-L3, L1-R1, L1-L2, L1-R3
inphase_comparasions = [1,4,5,7,10]
gen_phase_dist = []
gen_phase_stats = []

for g in range(2):
    # phase
    phase_dists = []
    nflies = len(np.unique(meta_matrix[g][:, 0]).astype(int).tolist()) # how many flies
    stat_mat = np.zeros((nflies, 2, len(rel_phase_comparasions))) # fly/stats/phase comparasion/

    # trial idxs
    tied_idxs = meta_matrix[g][:, 4] != 1 # presplit + postsplit - combined because no detectable adaptation
    
    p_cnt = -1
    for p in rel_phase_comparasions: # phases
        p_cnt += 1
        f_cnt = -1
        for fly in np.unique(meta_matrix[g][:, 0]).astype(int).tolist():
            f_cnt += 1
            fly_idxs = meta_matrix[g][:, 0] == fly # fly

            # find tied, LS, and RS data
            tied_trial_idxs = np.logical_and.reduce((tied_idxs, fly_idxs))

            # data
            tied_data = phase[g][p, :, tied_trial_idxs].flatten()

            # behavior classifier
            beh_tied = beh_class[g][tied_trial_idxs, :].flatten()

            # tarsi classifier (just for split period)
            tarsi_class_tied = tarsi_class[g][tied_trial_idxs, :].flatten()

            # filter data based on behavior and tarsi position
            filt_tied_data = tied_data[np.logical_and.reduce((beh_tied == 1, tarsi_class_tied != 7))]

            # filter out non-vals = -1 in this case
            filt_tied_data = filt_tied_data[filt_tied_data != -1]

            # just in case, filter out nans
            filt_tied_data = filt_tied_data[~np.isnan(filt_tied_data)]

            # scale remaining phase data by 2pi to transform it into the unit circle
            filt_tied_data = filt_tied_data *(2*np.pi)

            # store fly stats
            if len(filt_tied_data) >= step_threshold:
                tied_mean, tied_dispersion = estimate_ang_stats(filt_tied_data)
                if p in inphase_comparasions:
                    stat_mat[f_cnt, 0, p_cnt] = tied_mean

                    # distribution
                    tied_dist = transform_half_polar(filt_tied_data)
                    tied_prob, tied_bin_centers = compute_inphase_histogram(transform_half_polar(filt_tied_data), Nbins)

                    # interpolate between these values to determine the probability at average phase
                    f = interpolate.interp1d(tied_bin_centers, tied_prob)
                    stat_mat[f_cnt, 1, p_cnt] =  f(tied_mean)
                else:
                    stat_mat[f_cnt, 0, p_cnt] = transform_polar(tied_mean)
                    tied_prob, tied_bin_centers = compute_histogram(filt_tied_data, Nbins)

                    # interpolate between these values to determine the probability at average phase
                    f = interpolate.interp1d(tied_bin_centers, tied_prob)
                    stat_mat[f_cnt, 1, p_cnt] =  f(transform_polar(tied_mean))

            else:
                stat_mat[f_cnt, 0, p_cnt] = np.nan
                stat_mat[f_cnt, 1, p_cnt] = np.nan


            # concatenate data across flies for each distrbution
            if f_cnt == 0:
                if p in inphase_comparasions:
                    cat_tied_data = transform_half_polar(filt_tied_data)
                else:
                    cat_tied_data = filt_tied_data
            else:
                if p in inphase_comparasions:
                    cat_tied_data = np.concatenate((cat_tied_data, transform_half_polar(filt_tied_data)))       
                else:                
                    cat_tied_data = np.concatenate((cat_tied_data, filt_tied_data))

        phase_dists.append(cat_tied_data)
        
    # genotype info
    gen_phase_dist.append(phase_dists)
    gen_phase_stats.append(stat_mat)
```

```python
def bootstrap_stats(n_bootstraps, pval_thresh, bon_correction, pop1, pop2, plotting, comparasion):
    print(comparasion)
    n_pop1 = len(pop1)
    n_pop2 = len(pop2)
    mean_pop_diff = np.mean(pop2) - np.mean(pop1) # BL
    pooled_pop = np.concatenate((pop1, pop2))

    # bootstraping 
    mean_stat_diff = np.zeros(n_bootstraps)
    for j in range(n_bootstraps):
        resample_pop1 = np.random.permutation(pooled_pop)[0:n_pop1] # randomly permute from a uniform distribution
        resample_pop2 = np.random.permutation(pooled_pop)[n_pop1::]
        mean_stat_diff[j] = np.mean(resample_pop2) - np.mean(resample_pop1)

    if plotting:
        plt.figure(1, figsize =[5,5])
        hist, bin_edges = np.histogram(mean_stat_diff, bins=30, density=True)
        plt.hist(mean_stat_diff, bins = 30, color='k', density = True)
        plt.plot(np.array([mean_pop_diff, mean_pop_diff]), np.array([0,max(hist)]), color ='r')

    # compute p-value
    if mean_pop_diff> 0:
        n_sig_samples = len(np.where(mean_stat_diff>=mean_pop_diff)[0])
    else:
        n_sig_samples = len(np.where(mean_stat_diff<=mean_pop_diff)[0])
    pvalue = n_sig_samples/n_bootstraps
    if pvalue == 0: # correct for the number of bootstraps
        pvalue = 1/n_bootstraps
    print('p value = ', pvalue)

    # significance threshold with bonferroni correction
    if pvalue <= (pval_thresh/bon_correction):
        print('Statistically Significant')
    else:
        print('Not Statistically Significant')
```
